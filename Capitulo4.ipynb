## 1. Introdução
Neste capítulo estudaremos uns dos problemas fundamentais da Física: o movimento. O ramo da Física encarregado de analisar o movimento se chama **mecânica**, a qual por sua vez está dividida em dois sub-ramos: a **cinemática**, objeto de nosso estudo neste capítulo, e a **dinâmica**, a qual estudaremos posteriormente.

A cinemática consiste em analisar o movimento dos corpos, mas sem se importar com quem gerou tal movimento nem o por quê, pois essas perguntas se enquadram no ramo da **dinâmica**, a qual envolve o estudo das forças e como elas geram acelerações. Sendo assim, a cinemática estuda como acontece o movimento depois dele já ter começado.

A Física que tratamos neste curso de Física Geral é a denominada como **mecânica clássica**, a qual lida com corpos que não se movem com velocidades próximas à velocidade da luz, em cujo caso aconteceria um fenômeno relativista, dentro do ramo da **mecânica relativista**, conhecido como dilação do tempo. Ou seja, a altas velocidades, próximas à velocidade da luz, o tempo dependeria do movimento relativo de um corpo frente a outro.

Este capítulo se intitula "Cinemática em uma dimensão" porque apenas estudaremos o movimento ao longo de uma linha reta. Aliás, o que estudaremos aqui será o chamado **Movimento Retilíneo** (M.R). É claro que esse não é o movimento mais geral que existe. De fato, o movimento mais geral é o movimento ao longo do espaço de 3 dimensões no qual moramos. Os movimentos ao longo de duas dimensões, como o tiro parabólico, também são muito importantes em Física. Porém, para ambos se torna necessário usar e entender os vetores. Os vetores, porém, serão estudados nos próximos capítulos. Entretanto, o foco deste livro é, literalmente, a Física Geral, sem entrar em todos os detalhes, neste caso, da Cinemática.

O movimento de um corpo no espaço pode ser dividido em dois movimentos independentes:
1. Um **movimento de translação** do chamado **centro de massas**. Isto é, um movimento de translação de um ponto muito particular, no qual se concentra a massa toda do corpo.
2. Um **movimento de rotação** do corpo ao redor de um eixo de rotação que passa pelo centro de massas.

Estes dois movimentos podem, e devem, ser estudados separadamente. Novamente, nosso propósito é estudar uma Física Geral, dentro da qual o movimento rotacional está fora do escopo. Nosso foco, portanto, será o estudo da chamada **cinemática translacional**, ou seja, o estudo do movimento de um ponto. Para tratar tal movimento devemos introduzir o conceito de **partícula**. Para nós todos os corpos serão partículas, no sentido em que essa é a aproximação física a um ponto matemático porque os pontos não rotam, apenas se transladam. Essa aproximação faz sentido quando o corpo, mesmo que seja um corpo extenso, é visto de longe. Aliás, os movimentos do objeto devem ser muito maiores comparado com suas dimensões.

Resumindo, portanto, todas as premissas anteriores, podemos dizer que o que será estudado neste capítulo seria algo que poderíamos denominar "Cinemática translacional clássica da partícula".

### 2.1 Tempo

De fato, na Cinemática clássica o movimento é entendido como uma sequência de fotogramas, como se de um filme se tratasse. O resto das grandezas físicas que definiremos a seguir, que atuaram como funções, serão dependentes do tempo.

O tempo se denota habitualmente pela letra $t$ minúscula e no Sistema Internacional de Unidades é medida em segundos ($s$).

### 2.2 Referencial
Para estudar o movimento em uma dimensão não é necessário usar os vetores. Entretanto, exceto o tempo, todas as grandezas físicas que iremos definir a seguir, realmente são vetores.

Devido a isso, se torna necessário usar o chamado **referencial**. Um referencial é uma **origem de coordenadas**, pois isso é o que em Matemáticas se denomina **base vetorial**. Como a própria palavra diz, uma base vetorial é a base, a partir da qual todos os vetores do espaço vectorial podem ser construídos. Uma base vetorial tem que ser formada pelos chamados **vetores unitários**, pois através de uma combinação linear dos mesmos geramos qualquer vetor possível. Entretanto, podemos por enquanto não aprofundar tanto nesse assunto. Com uma base vetorial que confirme para onde é o sentido positivo do eixo horizontal, habitualmente **eixo X**, e para onde é o sentido positivo do eixo vertical, habitualmente **eixo Y**, será suficiente por enquanto. Esse, portanto, será nosso referencial: um novo plano cartesiano, só que, desta vez, usado não para representar funções, se não para representar a chamada **trajetória**, ou seja, a coleção dos pontos do espaço por onde o corpo passa. E como apenas cabe estudarmos o movimento unidimensional, a pesar de que habitualmente colocaremos esses dois eixos **X** e **Y**, apenas usaremos um deles para o movimento.

```python
# importing the required module
import matplotlib.pyplot as plt

# naming the x axis
plt.xlabel('Eixo X (m)', loc='right')
# naming the y axis
plt.ylabel('Eixo Y (m)')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid(linestyle = '--', linewidth = 0.5)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

# function to show the plot
plt.show()
```

Um último detalhe é que, na hora de colocar o referencial, é habitual explicitar em cada eixo quais são as unidades usadas para medir. Como observamos, as unidades escolhidas para proximamente averiguar a trajetória do corpo serão os metros.

### 2.3 Posição
Uma vez definido o referencial podemos, passe a redundância, referenciar a **posição** do objeto respeito a tal referencial. Num espácio tridimensional como no qual moramos, usam-se 3 coordenadas para determinar a posição de um objeto, para as quais se reservam as letras $x$ para a primeira coordenada, tipicamente coordenada horizontal, $y$ para a segunda coordenada, tipicamente coordenada vertical, e $z$ para a terceira, tipicamente coordenada da profundidade.

Por exemplo, um corpo (sempre pontual, não esqueçam) pode estar em algum ponto de um plano (sem a terceira coordenada de profundidade), digamos na posição de 3 metros à direta da origem ($x=3$) e a 2 metros à baixo da origem ($y=-2$).

```python
# importing the required module
import matplotlib.pyplot as plt
# x axis values
x = [3]
# corresponding y axis values
y = [-2]

# plotting the points
plt.scatter(x, y, marker='o')

# naming the x axis
plt.xlabel('Eixo X (m)', loc='right')
# naming the y axis
plt.ylabel('Eixo Y (m)')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid(linestyle = '--', linewidth = 0.5)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

# function to show the plot
plt.show()
```

Como vemos, para acertar a posição de um objeto é necessário um referencial, pois é ele quem determina desde onde temos que começar a contar posições. Ter um referencial, portanto, é fundamental e sem ele o seguimento do movimento de um corpo se torna ambíguo. Todo problema que precisemos resolver sobre cinemática deve ter um referencial. 

Entretanto, o movimento não pode depender do observador. Isso quer dizer que diferentes referenciais, tipicamente podem ver trajetórias diferentes, mas devem concordar com o movimento. Ou seja, o movimento deve ser independente do referencial. Isso, por sua vez, implica que **o referencial é uma escolha**.

De fato, já que o ponto ($x$,$y$)=($3m$,$-2m$) está referenciado a dois eixos, sempre podemos escolher um novo referencial no qual apenas seja necessária uma coordenada para relatar a posição desse objeto:

```python
# importing the required module
import matplotlib.pyplot as plt
# x axis values
x = [3]
# corresponding y axis values
y = [-2]

# plotting the points
plt.scatter(x, y, marker='o')

# naming the x axis
#plt.xlabel('Eixo X (m)', loc='right')
# naming the y axis
#plt.ylabel('Eixo Y (m)')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)

plt.grid(linestyle = '--', linewidth = 0.5)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

xorigem_ver2 = [-2, 2]
yorigem_ver2 = [-6, 6]
plt.plot(xorigem_ver2, yorigem_ver2, marker='', color='blue', linestyle="--")

xorigem_hor2 = [-6, 6]
yorigem_hor2 = [4, -4]
plt.plot(xorigem_hor2, yorigem_hor2, marker='', color='blue', linestyle="--")

plt.subplot(111).annotate("Eixo Y(m)", xy=(3, 3), xytext=(-0.5, 6.2))
plt.subplot(111).annotate("Eixo X(m)", xy=(3, 3), xytext=(5.2, 0.2))
plt.subplot(111).annotate("Eixo X'(m)", xy=(3, 3), xytext=(5.2, -5))
plt.subplot(111).annotate("Eixo Y'(m)", xy=(3, 3), xytext=(2, 5))

# function to show the plot
plt.show()
```

sendo ($x'$,$y'$)=($3.6m$,$0$) as coordenadas da posição do corpo nesse novo referencial, pois nosso novo $x$, chamado $x'$ será o valor de hipotenusa do triângulo retângulo formado pela coordenada $y$ e $x$ anteriores:
> $x'=\sqrt{(y^2+x^2)}=\sqrt{(-2m)^2+(3m)^2}=\sqrt{13m^2}=3.6m$

Nesta discussão sobre o melhor escolha sobre o referencial emergiu uma constante do movimento importante: a posição inicial. A posição inicial é a posição do corpo no instante de zero segundos, que é quando iniciamos o relógio do movimento. Dai se deduz que o corpo pode estar em posições diferentes em tempos diferentes. Isso se resume a dizer que **a posição é função do tempo**:
> $x=f(t)$

Ou de outra maneira, como habitualmente se coloca em Física:
> $x=x(t)$

E isso, por sua vez, implica que a posição é uma função matemática como as que analisamos no capítulo anterior, e sob a qual podemos aplicar todas as ferramentas estudadas. Veja, entretanto, as únicas diferenças: em Matemáticas $x$ era a variável independente e $y$, a dependente, enquanto agora $t$ será a variável independente e $x$ ou $y$ as posições horizontal e vertical respectivamente, que são funções dependentes do tempo, e portanto, as variáveis dependentes.

### 2.4 Intervalo temporal
Na análise do movimento de um corpo às vezes se requer saber informação em instantes de tempo, mas outras vezes se requer em **intervalos de tempo**. Podemos então, através de uma operação matemática, construir este último em função do anterior. Definiremos, portanto, um intervalo temporal, lido como "delta de t", como:
> $\Delta t = t_f - t_i$

onde $t_f$ será o instante de tempo final e $t_i$ será o instante de tempo inicial.

### 2.5 Intervalo espacial ou deslocamento
De maneira análoga ao intervalo temporal, podemos definir um **intervalo espacial**, neste caso ao longo do eixo X, como:
> $\Delta x = x_f - x_i = x(t_f)-x(t_i)$

Isto é, uma posição num instante final menos uma posição num instante inicial. Assim, um intervalo espacial é uma diferença de posições ou, de um ponto de vista mais geral, um vetor deslocamento ou, diretamente, num movimento unidimensional, um **deslocamento**.

É importante não confundir deslocamento com **percurso**, ou **caminho percorrido**. Quando se trata de um movimento em várias dimensões, o percurso será a distância, tipicamente denominada de $s$, percorrida pelo objeto ao longo de sua trajetória, o qual não igual ao deslocamento, pois este é apenas uma diferença entre o ponto inicial e o ponto final do percurso, os quais realmente seriam vetores. Em movimento unidimensional, porém, o objeto não pode sair de uma linha reta e, sendo assim, deslocamento e percurso seriam equivalentes.

Outro detalhe importante é o seguinte. Agora que podemos construir intervalos espaciais, podemos entender a própria definição da posição como se também fosse um intervalo espacial. De fato, um intervalo espacial respeito da posição zero:
> $x = x - 0 = \Delta x$

Esse resultado pode parecer irrelevante, mas possui importantes consequências. De fato, devemos entender a posição de um corpo como se realmente estivesse atrelada à posição zero, dada pela origem de coordenadas, ou pelo referencial. Entretanto, um intervalo espacial, quando a subtração não for respeito de zero, não teria esse atrelo. Isso, por sua vez, implica que poderemos zerar posições à vontade, como a posição inicial, simplesmente colocando o referencial nessa posição, porém, não poderemos zerar à vontade o resto de grandezas físicas, as quais definiremos a seguir, porque elas se constroem a partir do intervalo espacial. Ou dito de outra maneira, posições são parâmetros livres, pois sempre poderemos escolher seu valor, e seu sinal (positivo ou negativo), em função de onde for colocado o referencial. Já o resto de grandezas físicas não, e o único que pode mudar, em função do referencial escolhido, é o sinal, já que elas realmente são vetores e, desde esse ponto de vista, podem ir a favor ou em contra do que foi definido positivo pelo referencial.

## 2.6 Velocidade média
Logo depois de ter entendido o deslocamento, deveria surgir a pergunta de em quanto tempo foi feito esse deslocamento. Ou de outra maneira, quão rápido o corpo realizou esse deslocamento?

É assim como introduzimos novas grandezas físicas. Elas devem surgir para poder responder a perguntas necessárias sobre o movimento. A maneira mais lógica de definir a velocidade média é comparando um intervalo espacial com o intervalo temporal:
> $v_m = \frac{\Delta x}{\Delta t} = \frac{x(t_f)-x(t_i)}{t_f - t_i}$

É interessante reparar no procedimento para construir novas grandezas físicas: escolhemos grandezas físicas já definidas e realizamos operações matemáticas entre elas para criar outras novas que façam sentido.

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">
<strong>Dado o exemplo anterior, qual foi a velocidade média do corpo?</strong>

<p>Toda a informação necessária está no enunciado do exemplo. Assim, substituindo valores na definição e supondo que o instante de tempo inicial é zero, pois não temos outra informação:</p>

<p>

$v_m = \frac{x(t_f)-x(t_i)}{t_f - t_i} = \frac{(1m) - (5m)}{2s - 0s} = \frac{-4m}{2s} = -2 \frac{m}{s}$

</p>

<p>

A velocidade média do corpo é, por tanto, $2 m/s$ para o sentido negativo do eixo $X$. Ou seja, o corpo se move em sentido contrário ao que foi definido como positivo.

</p>

</div>

A velocidade média também é realmente um vetor. Entretanto, ao igual que acontece com o deslocamento, em movimentos ao longo de uma dimensão, o carácter vetorial da velocidade média simplifica a um número, que realmente indicaria o tamanho do vetor, e a um sinal, que indicaria o sentido do vetor.

Como a posição de um objeto em movimento é uma função do tempo, a qual podemos representar num gráfico frente ao tempo, do ponto de vista geométrico, a velocidade média representaria a inclinação da reta que une os pontos $x(t_i)$ com $x(t_f)$. Podemos visualizar isso através de um exemplo.

<strong>A posição de um corpo muda com o tempo segundo a função $x(t)=(2\frac{m}{s^2})t^2-(1m)$. (a) Calcule a posição no instante de tempo $2s$. (b) Calcule o deslocamento entre os instantes $0s$ e $2s$. (c) Calcule sua velocidade média entre $0s$ e $2s$. (d) Levante o gráfico da função posição junto com a reta, cuja inclinação representa à velocidade média. (e) Desenhe a trajetória do movimento.</strong>

Antes de começar é importante reparar em algo. Quando estudávamos Matemática, uma função como a que aparece nesta questão seria escrita simplesmente assim:

$f(x) = 2x^2-1$

Entretanto, em Física, como cada variável representa uma grandeza física, o $x$ de Matemáticas será o tempo e o $f(x)$ será a posição, a qual é mesmo função do tempo:

$x(t) = 2t^2-1$

Além disso, há outra diferença. Em Física os coeficientes dos polinômios têm sentido físico mesmo e, sendo assim, eles têm unidades. E as unidades que tais coeficientes devem ter serão aquilas que deixam a equação homogênea, no sentido em que, depois de substituir, as unidades devem ser as mesmas à esquerda que à direita do símbolo =.

Quando um problema de cinemática como este fornece a função posição, podemos conseguir facilmente qualquer informação a partir da mesma.

## 2.7 Velocidade instantânea
No cálculo da velocidade média obtemos uma informação, como a palavra diz, média do que acontece no intervalo temporal escolhido. Entretanto, ao longo de esse intervalo de tempo, o corpo poderia ter velocidades diferentes em cada instante. É por isso que precisamos definir uma nova grandeza física que não forneça informação de intervalos, mas do que acontece em instantes de tempo.

Podemos construir a velocidade instantânea a partir da velocidade média, tomando um intervalo temporal infinitamente estreito. Mas essa operação é conhecida, pois equivale a tomar a derivada da função posição respeito do tempo:
> $v=\lim_{\Delta t\to 0} v_m = \lim_{\Delta t\to 0} \frac{\Delta x}{\Delta t} = \frac{d}{dt}x$

Observe como, para calcular a velocidade instantânea, precisamos derivar a função posição respeito do tempo e o resultado disso será uma nova função, que também depende do tempo e, portanto:
> $v=v(t)$

Veja que essa é uma diferença fundamental entre a velocidade média a instantânea. A primeira é sempre um número, ou uma constante, pois equivale a uma inclinação. Já a segunda é mesmo uma função, a qual, se torna um número quando na própria função é insertado o instante de tempo no qual queremos calcular a velocidade instantânea.

Efetivamente, a velocidade média e a velocidade instantânea são grandezas físicas com significados diferentes e a segunda é bem mais usada que a primeira. É por causa disso que, por abuso de linguagem, na hora de enunciar a velocidade instantânea falaremos diretamente de **velocidade**.

## 2.7 Intervalo de velocidades
Da mesma maneira que definimos intervalos temporais e intervalos espaciais, podemos também definir um **intervalo de velocidades**, sendo este construido da seguiente maneira:
> $\Delta v = v_f - v_i = v(t_f) - v(t_i)$

O intervalo de velocidade não tem muita utilidade prática, mas serve para definir as duas grandezas físicas que faltam.

## 2.8. Aceleração média
De maneira análoga à velocidade média, podemos definir uma **aceleração média**:
> $a_m = \frac{\Delta v}{\Delta t} = \frac{v(t_f) - v(t_i)}{t_f-t_i}$

A aceleração média é construída como um quociente entre velocidades, cujas unidades no S.I. são $m/s$, e tempos, cujas unidades no S.I. são segundos. Sendo assim, as unidades da aceleração média, serão $m/s$ divididos por $s$, ou seja, $m/s^2$.

## 2.9 Aceleração instantânea
Igual que acontece com a velocidade média, a aceleração média apenas fornece informação do que acontece dentro do intervalo temporal escolhido. É por isso que, precisamos definir uma nova aceleração que informe do que acontece em instantes. Para isso, procedemos de maneira análoga a como foi feito quando definimos a velocidade instantânea:
> $a = \lim_{\Delta t\to 0} a_m = \lim_{\Delta t\to 0} \frac{\Delta v}{\Delta t} = \frac{d}{dt}v$

Isto é, a aceleração instantânea e uma função (também dependente do tempo) construída a partir da derivada da função velocidade instantânea.

Podemos continuar o desenvolvimento de cima insertando a definição da velocidade instantânea:
> $a = \frac{d}{dt}v = \frac{d}{dt}\frac{d}{dt}x = (\frac{d}{dt})^2 v = \frac{d^2}{dt^2} x$

Como vemos, podemos entender a aceleração instantânea não só como a (primeira) derivada da velocidade, mas também como a segunda derivada da posição. Ou dito de outra maneira, e também de maneira geral, a partir da função posição podemos deduzir todas as grandezas físicas que foram definidas até aqui.

Ao igual que acontecia com a velocidade instantânea, na hora de enunciar a aceleração instantânea, diremos diretamente que ela é a **aceleração**.

## 3. Movimento retilíneo com aceleração constante
No exemplo anterior, vimos como deduzir a função velocidade a partir da função posição, derivando esta duas vezes. O resultado foi uma aceleração constante, i.e., uma aceleração cujo valor não muda no tempo. Isso apenas foi assim porque a função posição do exemplo foi um polinômio de grau 2, mas de maneira geral a aceleração é uma função variável. A grande maioria de movimentos possuem realmente uma aceleração variável, os quais, no caso de uma dimensão, são tipicamente chamados de **Movimentos Retilíneos Uniformemente Variados (M.R.U.V)**. Entretanto, os movimentos com a aceleração constante, também chamados em uma dimensão como **Movimento Retilíneo Uniformemente Acelerado (M.R.U.A.)**, representam um tipo de movimento simples, o qual é tipicamente usado em Física, como aproximação de outro tipo de movimentos com aceleração variável como a **queda livre** dos corpos.

A queda livre unidimensional é um tipo particular de movimento, ao qual estamos muito habituados, que consiste simplesmente em largar um corpo desde uma certa altura, ou arremesar o mesmo para cima ou para abaixo com alguma velocidade inicial. Durante a queda livre todos os corpos estão submetidos à **aceleração da gravidade** ($g$), a qual é tratada como constante, de valor $g=9.8m/s^2$, a qual aqui aproximaremos por $g\approx 10m/s^2$. Insistimos em que, realmente, a queda livre é um movimento com aceleração variável, já que a aceleração da gravidade, segundo a Lei de Gravitação Universal, depende do quadrado da distância do objeto até a centro da Terra. Entretanto, para objetos que se movem poucos metros no entorno da superfície, podemos aceitar tal aproximação.

### 3.1 Equação da velocidade como função do tempo
Podemos deduzir a equação da velocidade para um movimento com aceleração constante usando a própria definição da aceleração e transformando essa equação, uma equação diferencial, numa equação integral. Isso é necessário porque, desta vez, precisamos de uma equação que forneça informação não do que aconteça em instantes, mas sim do que aconteça em intervalos.

> $a=\frac{dv}{dt} \Longleftrightarrow dv = a dt \Longleftrightarrow \int_{v_i}^{v_f}dv = \int_{t_i}^{t_f} a dt \Longleftrightarrow \big[v\big]_{v_i}^{v_f} = a \big[t \big]_{t_i}^{t_f}$

A seguir substituímos os limites nos extremos dos intervalos e simplificamos os subscritos. Usaremos que:
- o instante de tempo inicial ($t_i$) será zero.
- o instante de tempo final ($t_f$) será livre, passando a ser diretamente $t$.
- a velocidade inicial ($v_i$) será chamada $v_0$, ou seja, a velocidade do corpo quando o tempo é zero.
- a velocidade final ($v_f$) também será livre, passando a ser diretamente $v$.

> $v_f - v_i = a(t_f - t_i) \Longleftrightarrow v - v_0 = at$

Ou, finalmente:
> |  $v = v_0 + at$  |
> | :---: |

Esta é a equação da velocidade para um movimento com aceleração constante. Esta é uma equação que depende linearmente do tempo, na qual tanto $v_0$ quanto $a$ são constantes do movimento.

### 3.2 Equação da posição como função do tempo
Podemos também deduzir a equação da posição para um movimento retilíneo com aceleração constante usando a equação anterior e a definição da velocidade. O procedimento seria análogo: transformar uma equação diferencial em uma equação integral.
> $v = v_0 + at \Longleftrightarrow \frac{dx}{dt} = v_0 + at \Longleftrightarrow dx = (v_0 + at)dt \Longleftrightarrow \int_{x_i}^{x_f}dx = \int_{t_i}^{t_f}v_0dt + \int_{t_i}^{t_f}atdt \Longleftrightarrow \big[x \big]_{x_i}^{x_f} = v_0 \big[t \big]_{t_i}^{t_f} + a\big[\frac{t^{2}}{2} \big]_{t_i}^{t_f}$

Igual que antes, substituímos os limites nos extremos dos intervalos e simplificamos os subscritos, segundo o seguinte critério:
- o instante de tempo inicial ($t_i$) será zero.
- o instante de tempo final ($t_f$) será livre, passando a ser diretamente $t$.
- a posição inicial ($x_i$) será chamada $x_0$, ou seja, a posição do corpo quando o tempo é zero.
- a posição final ($x_f$) também será livre, passando a ser diretamente $x$.

> $x_f - x_i = v_0 (t_f - t_i) + \frac{a}{2}(t_f^2 - t_i^2) \Longleftrightarrow x - x_0 = v_0 (t-0) + \frac{1}{2}a(t^2 - 0^2)$

E finalmente, simplificando:
> |  $x = x_0 + v_0 t + \frac{1}{2}at^2$  |
> | :---: |

Esta é a equação da posição para um movimento com aceleração constante. Esta é uma equação que depende quadraticamente do tempo, e onde $x_0$, $v_0$ e $a$ são constantes do movimento.

Desta maneira, juntando as 3 equações para um movimento com aceleração constante, e enfatizando que tanto $x$, quanto $v$, quanto $a$ são funções do tempo, teríamos que:

> |  $x(t) = x_0 + v_0 t + \frac{1}{2}at^2$  |
> | :---: |
> |  $v(t) = v_0 + at$  |
> | $a(t) = a$ |

onde a última equação é uma das chamadas **equações triviais**, as quais não aportam informação alguma. Portanto, eliminando esta última, teríamos um sistema de 2 equações **linearmente independentes** com, tipicamente, 3 incógnitas ($x$, $v$ e $t$), supondo que o resto das constantes ($x_0$, $v_0$ e $a$) sejam dados. E sendo assim, para que o sistema tenha solução, apenas basta com que 1 dessas 3 incógnitas sejam um dado. 

Quando, no item 2.3, falávamos sobre os referenciais explicamos como eles provêm do que em Matemática se denomina base vetorial. O conceito de base de um espaço multidimensional é muito amplo, pois ele pode ser aplicado não só aos vetores, mas também a outros entes matemáticos como os polinômios. De fato, também é possível definir uma base para o chamado espaço dos polinômios. Essa base, no caso de usar mesmo o tempo como variável independente, estaria composta pelas seguintes componentes:
> ${1, t, t^2, t^3, t^4, ...}$

já que a partir de uma combinação linear dessas componentes poderíamos construir qualquer polinômio. De fato, no que tange a nosso caso, a partir dessa base polinomial, podemos construir nossas funções, ou polinômios, $x(t)$ e $v(t)$:

$
\begin{pmatrix} 
   x_0 & v_0 & \frac{1}{2}a\\

\end{pmatrix}
\begin{pmatrix} 
   1 \\ t \\ t^2 
\end{pmatrix}=
\begin{pmatrix}
   x(t) \\ v(t)
\end{pmatrix}
$

## 4. Movimento retilíneo e uniforme (M.R.U.)
No exemplo anterior, vimos como deduzir a função
