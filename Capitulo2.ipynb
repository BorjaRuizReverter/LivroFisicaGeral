## 1. Introdução
No capítulo anterior estudamos como a Física é construída a partir das grandezas físicas e, sobre todo, a partir das relações entre elas. Efetivamente, cada área da Física, emprega um conjunto específico de grandezas físicas e, a partir desse conjunto, podemos expandir o conhecimento construindo relações entre as mesmas que façam sentido, ou que satisfaçam a lógica.

De fato, proximamente estudaremos relações entre grandezas físicas através de equações como essa: $v = v_0 + a \cdot t $, a qual relaciona a velocidade $v$ de um corpo em um determinado instante de tempo $t$ com outras grandezas físicas como sua aceleração $a$ ou sua velocidade inicial $v_0$, ou relações como essa: $ F = m \cdot a $, na qual a força $F$ aplicada a um corpo de massa $m$ vai lhe fornecer uma aceleração de valor $a#.

Como vemos, as grandezas físicas se relacionam entre elas. Ou dito de outra maneira mais técnica, entre as grandezas físicas emergem **relações funcionais**. Ou seja, uma grandeza física pode ser **função** de outra ou de outras. Especificamente, se sabemos que $v = v_0 + a \cdot t $, isso implica que a velocidade do corpo depende do tempo, já que nessa equação, tanto $v_0$ quanto $a$ são constantes do movimento, ou seja, números fixos que não variam. Por isso, a relação acima pode ser também expressa como $ v = v(t) = v_0 + a \cdot t $, para assim explicitar que, efetivamente, a velocidade é função do tempo.

Neste capítulo estudaremos do ponto de vista matemático as funções com o objetivo de entender, do ponto de vista físico, como são as relações entre as grandezas físicas implicadas. Contudo, dentro da linguagem da matemática não falaremos de grandezas físicas, se não de variáveis, as quais denotaremos tipicamente por letras do final do abecedário como $x$, $y$ ou $z$, enquanto que as constantes serão denotadas por letras do começo como $a$, $b$, $c$, etc... Também, ao contrário de em Física, não diremos que certa grandeza física como a velocidade é função do tempo, ou seja, $v = v(t)$, se não que certa variável $y$ é função de $x$, o qual tipicamente é escrito como $y=f(x)$.

## 2. O que é uma função?

Em Matemática, uma função é basicamente uma **transformação**. Isso quer dizer que cada input, ou valor de entrada, chamado x, transforma-se numa saída, ou output, chamado y, através da função. Tal esquema seria algo assim:
> $x \hspace{0.2cm} \underrightarrow{f} \hspace{0.2cm} y$

ou também pode ser assim, para insistir em que transformamos conjuntos de inputs em conjuntos de ouputs:
> $\{x_1, x_2, x_3, ...\} \hspace{0.2cm} \underrightarrow{f} \hspace{0.2cm} \{y_1, y_2, y_3, ...\}$

É importante notar que cada $x$ de entrada terá um $y$ de saída, o qual, casualmente pode ainda ser o mesmo. Mas nunca dois $x$ diferentes desembocarão no mesmo $y$. Isto é porque uma função não pode ser ambígua. Ou seja, uma função deve ser **unívoca**.

De maneira geral, a notação esquemática anterior se resume simplesmente a escrever:
>$y = f(x)$

que é lido como "$y$ é função de $x$".

Isso implica que existe uma ordem ou hierarquia, pois é como variável $x$ com a qual fazemos algo para obter $y$ e não ao contrário. Ou dito de outra maneira, a variável $y$ depende de qual seja $x$. É por isso que afirmamos que $x$ é a **variável independente**, já que somos livres na hora de escolher ela, mas $y$ se diz que é a **variável dependente** porque seu valor está atrelado ao valor de $x$ e a como a função diga como realizar tal transformação.

Um exemplo de função simples seria $f(x) = 2x +1$. Neste caso, todo número de entrada $x$ é multiplicado por 2 e, depois, a isso é somado 1. Esse resultado de saída será nosso $y$.

Uma função, portanto, é uma transformação de um conjunto de valores de entrada, chamado **domínio da função**, em outro conjunto de valores de saída, chamado **imagem da função**. Visto dessa maneira, falando de conjuntos, ao invés de valores individuais seria algo assim:
> $dominio \hspace{0.2cm} \underrightarrow{f} \hspace{0.2cm} imagem$

O domínio ou a imagem de uma função tipicamente consiste no conjunto dos números reais $\mathcal{R}$, mas às vezes pode estar formado por apenas um subconjunto do mesmo, ou ainda o conjunto dos números complexos $\mathcal{C}$ ou um subconjunto dele.

</p>
<p>

Respeito à imagem da função, no caso da função $f(x) = 2x +1$, observamos que podemos obter qualquer número $y$ como resultado de saída e, portanto, a imagem dessa função será o conjunto dos números reais. No caso da função logaritmo, podemos perceber que a imagem da função também será o conjunto dos números reais. Entenderemos isso mais na frente!

## 3. Representando uma função

### 3.1 Plano cartesiano

Para representar uma função usamos o chamado **plano cartesiano**, que é a ferramenta usada para representar uma função de uma variável em uma variável. Tal transformação gera pares de pontos ($x$,$y$) chamado **pares ordenados**, os quais se representam em tal plano. O plano cartesiano se construí de maneira que a coordenada independente vai no chamado **eixo de abscissas**, ou **eixo** $x$, e a coordenada dependente no chamado **eixo de ordenadas**, ou **eixo** $y$.


```python
# importing the required module
import matplotlib.pyplot as plt
# x axis values
x = [-1,0,1,2]
# corresponding y axis values
y = [-1,1,3,5]

# plotting the points
plt.scatter(x, y, marker='')

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid(linestyle = '--', linewidth = 0.5)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

# function to show the plot
plt.show()
```

O plano cartesiano pode ser separado em 4 **quadrantes**, a contar desde o primeiro situado nos valores positivos tanto de $x$ quanto de $y$ até o quarto de maneira contrária ao movimento das agulhas de um relógio. Isso é assim porque, como veremos, os ângulos começam a se contar a partir de chamado **semieixo positivo** de $x$ e na direção contrária ao relógio.

```python
# importing the required module
import matplotlib.pyplot as plt
# x axis values
x = [-1,0,1,2]
# corresponding y axis values
y = [-1,1,3,5]

# plotting the points
plt.scatter(x, y, marker='')

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid(linestyle = '--', linewidth = 0.5)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

#adding text inside the plot
plt.text(2, 3, '1$^{er}$ quadrante', fontsize = 12)
plt.text(-4, 3, '2° quadrante', fontsize = 12)
plt.text(-4, -3, '3$^{er}$ quadrante', fontsize = 12)
plt.text(2, -3, '4° quadrante', fontsize = 12)

# function to show the plot
plt.show()
```

Mas também pode ser visualizado como estivesse dividido por uma linha bissetriz que corta os quadrantes diagonalmente de esquerda para direita. Tais linhas se conhecem como **bissetriz dos quadrantes ímpares** ou **bissetriz dos quadrantes pares**.

```python
# importing the required module
import matplotlib.pyplot as plt
# x axis values
x = [-1,0,1,2]
# corresponding y axis values
y = [-1,1,3,5]

# plotting the points
plt.scatter(x, y, marker='')

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid(linestyle = '--', linewidth = 0.5)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

xorigem_ver = [-6, 6]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='blue', linestyle="--")

xorigem_ver = [-6, 6]
yorigem_ver = [6, -6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='blue', linestyle="--")

#adding text inside the plot
plt.text(0.5, 3, 'Bissetriz dos quadrantes', fontsize = 12)
plt.text(0.5, 2, 'ímpares', fontsize = 12)
plt.text(0.5, -3, 'Bissetriz dos quadrantes', fontsize = 12)
plt.text(0.5, -4, 'pares', fontsize = 12)

# function to show the plot
plt.show()
```

## 4. Tipos de função
Dentro do conjunto das funções analíticas, ou seja, aquelas que podem ser escritas, existe uma enorme variedade delas. Veremos a seguir algumas das principais. 

### 4.1 Função polinômio
A função polinômio, ou função polinômica, está construída por vários "nômios". Ou seja, junções de uma constante com uma variável elevado a algum número inteiro, as quais somam uma com as outras. De maneira geral, tal função tem esta forma:
> $f(x) = a_n \cdot x^n + a_{n-1} \cdot x^{n-1} + a_{n-2} \cdot x^{n-2} + ... + a_{0} \cdot x^0$

onde $a_n$, $a_{n-1}$, $a_{n-2}$,... são coeficientes quaisquer.

Um exemplo de função polinômica seria a **função linear**. Para isso, fazemos acima $n=1$. Desta maneira,
> $f(x) = a_1 \cdot x + a_0$

Ou escrita de uma maneira mais comum,
> $f(x) = a \cdot x + b$

onde $a$ e $b$ são os chamados coeficientes da reta. Aliás, o coeficiente angular e o coeficiente linear respectivamente.

Outro exemplo de função polinômica é **função identidade**, a qual pode ser vista como um caso particular da função linear. De fato, fazendo $a=1$ e $b=0$ obtemos tal função:
> $f(x) = x$

As funções quadráticas ou cúbicas são também exemplos de funções polinomiais. A função quadrática ou função parabólica teria esta forma:
> $f(x) = a \cdot x^2 + b \cdot x + c$

e a cúbica esta outra:
> $f(x) = a \cdot x^3 + b \cdot x^2 + c \cdot x + d$

De maneira geral, as funções polinômicas são particularmente importantes porque, no entorno de um punto, podemos usá-la para obter aproximadamente qualquer função. Ou dito de outro modo, qualquer função analítica pode ser aproximada em função de polinômios.

### 4.2 Funções logarítmicas
O logaritmo é uma operação matemática muito importante porque permite entender a operação de multiplicar como uma soma e a operação de divisão como uma subtração. Uma consequência disso é que o logaritmo pode ser usado para analisar números que diferem muitas ordens de grandeza.

Na hora de usar o logaritmo temos que determinar a chamada base, que tem a ver com quão rápido analisamos o crescimento de uma função. Por ser a base decimal, é muito comum usar a base 10, em cujo caso escriviremos $log_{10}$ ou diretamente $log$. Assim, podemos aplicar a operação logaritmo em base 10 a uma certa função qualquer:
> $log f(x)$

Por exemplo, podemos aplicar o logaritmo a nossa função linear anterior, obtendo assim uma função logarítmica:
> $f(x) = log (2x+1)$

Também, na hora de operar com logaritmos, é muito comum usar a chamada base $e$, onde $e$ é o chamado número de Euler: um número de valor 2.718..., o qual emerge naturalmente como um limite de sucessões. No caso de aplicar tal operação escreveremos $log_e$ ou diretamente $ln$, o qual significa logaritmo natural, pois o logaritmo natural e função exponencial, que veremos a seguir, emerge com frequência em padrões vistos na Natureza.

Assim, podemos igualmente fazer todo o anterior: aplicar nosso logaritmo natural a qualquer função, como pode ser uma função linear qualquer, para obter uma função logarítmica natural.
> $f(x) = ln (2x+1)$

### 4.3 Funções exponenciais
As funções exponenciais são aquilas envolvendo o número de Euler, sendo a função abaixo a mais simples que pode ser construída:
> $f(x) = e^x$

A função exponencial também é muito importante, pois vários fenômenos no ramo da Biologia como o crescimento de uma nova espécie melhor adaptada ou espalhamento de um vírus segue esse tipo de função. Além disso, a função exponencial surge como função inversa à função logaritmo.

Repare-se como a função exponencial pode ser aplicada a qualquer outra função. Desta maneira, igual que antes, podemos aplicar a função exponencial a nossa função linear para construir uma nova função como:
> $f(x) = e^{2x+1}$

### 4.4 Funções trigonométricas
As funções trigonométricas surgem em vários ramos da matemática como a geometria analítica ou a álgebra vetorial. Existem varias funções trigonométricas, sendo a função seno e cosseno as principais, as quais se denotam por:
> $f(x) = sen(x)$
> $f(x) = cos(x)$

O resto de funções trigonométricas conhecidas como tangente, secante, arco secante, arco seno e arco cosseno podem ser construídas a partir destas duas.

## 5. Características de uma função
### 5.1 Pontos de corte
Na hora de representar uma função, podemos observar que alguns pontos são mais importantes que outros. Dentre esses valores estão os **pontos de corte da função com o eixo de abcissas** e o **ponto de corte da função com o eixo de ordenadas**.

O primeiro deles é particularmente importante. Tanto que os pontos de corte da função com o eixo de abcissas se denominam **soluções da função** ou **raízes da função**, pois é nesses valores de $x$ onde a função se anula.

Para entender a importância dessas soluções ou raízes da função, aplicado à Física, calcular a raízes de uma função qualquer onde, por exemplo, a velocidade dependa do tempo, equivale a saber em que instante de tempo o corpo para.

Para calcular tais raízes, o único que temos que fazer é impor a condição de que a função seja zero. Ou seja, fazer com que:
> $f(x) = 0$ 

Desta maneira, ao fazer isso, a função se torna uma equação, de onde só restaria isolar a incógnita. O valor dessa incógnita nos dirá onde a função se anula.

### 5.4 Paridade de uma função
Na hora de representar uma função, um detalhe muito importante é sua simetria respeito dos eixos. Tal aspecto é importante porque, se uma função é simétrica, sua representação num quadrante do plano cartesiano pode ser deduzida a partir da forma da função em outro quadrante e, portanto, podemos representar a função rapidamente.

Uma função se diz ser **par** quando é satisfeito que:
> $f(-x) = f(x)$

Ou seja, que a função devolva o mesmo valor tanto para um $x$ particular quanto para seu oposto $-x$. Se uma função é par haverá uma simetria especular da função respeito do eixo de ordenadas. Ou, de outro ponto de vista, haverá uma simetria rotacional respeito do eixo de ordenadas. Aliás, tomando como eixo de rotação o eixo de ordenadas, obteremos a mesma função ao rotar ela 180° respeito desse eixo.

Por outro lado, diz-se que uma função é **ímpar** quando:
> $f(-x) = - f(x)$

Ou seja, que a função original se torne negativa quando o valor $-x$ é insertado na função. Se uma função é ímpar haverá uma simetria especular da função respeito à bissetriz dos quadrantes pares ou, de outro ponto de vista, haverá uma simetria rotacional respeito da origem do plano cartesiano. Aliás, tomando como eixo de rotação a origem do plano cartesiano, obteremos a mesma função ao rotar ela 180° respeito desse eixo.

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>Estude a paridade das funções $f(x)=2x+1$ e $f(x)=ln(x)$.</b>

<p>

Nenhuma das funções anteriores satisfazem tais condições de paridade. De fato, a função linear $f(x)=2x+1$ não satisfaz a condição de função par porque ao insertar $-x$ no lugar de $x$ o resultado seria uma nova função diferente da original: $f(-x)=2(-x)+1=-2x+1$. Uma olhada no seu próprio gráfico seria suficiente para ver que o eixo de ordenadas não atua como se fosse um espelho. Também não há simetria ímpar, pois $f(-x)=-2x+1\neq - (2x+1) = -f(x).$

</p>
<p>

De maneira geral, as funções lineares não têm como ter simetria par, já que realmente elas são polinômios de grau 1. E será assim para qualquer polinômico de grau impar. Já os polinômios de grau par podem ter simetria par desde que todos seus nômios sejam de grau par. Por exemplo, podemos comprovar que funções como $f(x)=x^2+1$ ou $f(x)=x^4+x^2$ satisfazem a condição de simetria par. Já todo o dito seria equivalente para os polinômios de grau impar, com todos seus nômios de grau ímpar, respeito da simetria ímpar.

</p>
<p>

De maneira análoga, a função $f(x)=ln(x)$ também não guarda simetria par nem impar, pois a função não existe para valores negativos de $x$.

</p>
</div>

É importante salientar que algumas funções trigonométricas como aquilas envolvendo o cosseno teriam simetria par já que o cosseno se caracteriza por que $cos(x) = cos(-x)$. Em câmbio, outras funções envolvendo a função seno, poderiam ter simetria ímpar, já que o seno se caracteriza por que $sen(-x)=-sen(x)$.


### 5.5 Tendência de uma função

A tabela valores é uma ferramenta muito útil para representar uma função, mas não é panaceia. É importante reparar que com a tabela de valores registramos alguns poucos pontos por onde passa a função, porém, tal varredura pode se tornar insuficiente quando a função é mais complexa. 

Uma maneira de ter certeza sobre por onde passa uma função ou para onde ela se dirige é usando os chamados **limites da função**. O limite de uma função é uma operação matemática para estudar para onde tende a mesma quando os valores de $x$ que ela usa como input tendam a um valor particular $x_0$. Todo o qual se denota assim:

> $\lim_{x\to x_0} f(x)$

Para entender como usar os limites e seus benefícios podemos usar a função clássica $f(x)=1/x$, a qual representa o paradigma para o estudo dos mesmos.

Um levantamento de uma tabela de valores a partir da tal função usando uns poucos valores de $x$ a ambos os lados da reta real nos daria o seguinte:

|  x  |  y  |
|:--: |:---: |
| -2 | -0,5 |
| -1 | -1 |
| 0 | $\nexists$ |
| 1 | 1 |
| 2 | 0,5 |

Entretanto, ao colocar esses pontos no plano cartesiano, não conseguimos imaginar qual seria a forma da função.

```python
# importing the required module
import matplotlib.pyplot as plt

# x axis values
x = [-2, -1, 1, 2]
# corresponding y axis values
y = [-0.5, -1, 1, 0.5]

# plotting the points
plt.scatter(x, y, marker='o')

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

plt.grid(linestyle = '--', linewidth = 0.5)

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

# function to show the plot
plt.show()
```

Como vemos uma tabela de valores pode se tornar insuficiente para representar uma função. Uma saída é continuar levantando mais pontos. Entretanto, essa tarefa pode ser tornar demorada. Por isso, pode ser conveniente usar outra ferramenta mais rápida como precisamente o estudo dos limites da função quando a variável $x$ tende a determinados pontos. Esse aqui é um desses casos. Estudemos então para onde tende a função quando os valores de $x$ tendem a pontos-chave, começando por $x=0$.

Se insertamos o valor de $x=0$ na função e usamos qualquer calculadora observaremos que a mesma nos devolve um valor que não existe porque dividir por zero no está definido. Entretanto, continuamos com a dúvida de quanto vale a função em pontos próximos a zero.

Uma estratégia seria não insertar o valor $x=0$ diretamente, mas sim valores que se aproximem a ele. Podemos desta maneira seguir insertando pontos na tabela de valores anterior:

|  x  |  y  |
|:--: |:---: |
| 0 | $\nexists$ |
| 0.001 | 1000 |
| 0.01 | 100 |
| 0.1 | 10 |
| 1 | 1 |

Observe como, começando por $x=1$ e tomando valores cada vez menores, nos aproximando a zero, os valores da função aumentam muito. Desta maneira, continuando com essa mesma progressão, poderíamos confirmar que quando os valores de $x$ tendem a zero a função se dispara até o infinito. Ou seja, quando nos aproximamos a zero, vindo da direita ($x\to 0^{+}$), a função tende a $\infty$.
> $\lim_{x\to 0^{+}} f(x) = \lim_{x\to 0^{+}} \frac{1}{x} = \frac{1}{0^{+}} = \infty$

Podemos fazer essa mesma análise, nos aproximando a zero, mas desde o lado negativo. Levantemos os valores para alguns desses pontos:

|  x  |  y  |
|:--: |:---: |
| -1 | -1 |
| -0.1 | -10 |
| -0.01 | -100 |
| -0.001 | -1000 |
| 0 | $\nexists$ |

Aliás, a medida que me aproximo a $x=0$, vindo do lado esquerdo ($x\to 0^{-}$), a função se dispara, mas o faz para os números negativos, ou seja, a função tende a $-\infty$.
> $\lim_{x\to 0^{-}} f(x) = \lim_{x\to 0^{-}} \frac{1}{x} = \frac{1}{0^{-}} = -\infty$

Com tais novas informações ao respeito de para onde tende a função quando nos aproximamos a $x$ podemos levantar um novo gráfico. Observemos como a função nunca atinge o valor zero. Isso acontece porque em $f(x=0)$ há uma descontinuidade.

```python
# importing the required module
import matplotlib.pyplot as plt

# x axis values
x_left = [-2, -1, -0.5, -0.25, -0.1, -0.01, -0.001]
# corresponding y axis values
y_left = [-0.5, -1, -2, -4, -10, -100, -1000]


# x axis values
x_right = [0.01, 0.1, 0.25, 0.5, 1, 2]
# corresponding y axis values
y_right = [100, 10, 4, 2, 1, 0.5]

# plotting the points
plt.plot(x_left, y_left, marker='o')

# plotting the points
plt.plot(x_right, y_right, marker='o', color='#1f77b4')

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

plt.grid(linestyle = '--', linewidth = 0.5)

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

# function to show the plot
plt.show()
```

Como vemos, usando os limites, temos conseguido averiguar que acontece com a função $f(x)=1/x$, quando $x$ se aproxima a zero. Grande passo. Mas ainda resta averiguar que acontece quando os valores de $x$ são grandes, tanto positivamente quanto negativamente. Podemos usar para isso uma abordagem similar no sentido de ir dando valores de $x$ cada vez maiores. Porém, desta vez vamos fazer diretametamente as contas usando os limites, pois esse deve ser o procedimento visto que tudo é entendido. Começando então por valores de $x$ grandes e positivos, teríamos que:
> $\lim_{x\to \infty} f(x) = \lim_{x\to \infty} \frac{1}{x} = \frac{1}{\infty} = 0$

Ou seja, a função tende a zero a medida que os valores de $x$ aumentam.

Por outro lado, usando os valores negativos de $x$, teriamos que:
> $\lim_{x\to -\infty} f(x) = \lim_{x\to -\infty} \frac{1}{x} = \frac{1}{-\infty} = 0$

Ou seja, a função também tende a zero a medida que os valores de $x$ aumentam negativamente.

Com essa nova informação, agora sim, podemos levantar o gráfico para o todo o domínio da função, excluindo $x=0$, o qual não faz parte desses domínio.


```python
# importing the required module
import matplotlib.pyplot as plt

# x axis values
x_left = [-6, -4, -3, -2, -1, -0.5, -0.25, -0.1, -0.01, -0.001]
# corresponding y axis values
y_left = [-0.167, -0.25, -0.33, -0.5, -1, -2, -4, -10, -100, -1000]


# x axis values
x_right = [0.01, 0.1, 0.25, 0.5, 1, 2, 3, 4, 6]
# corresponding y axis values
y_right = [100, 10, 4, 2, 1, 0.5, 0.33, 0.25, 0.167]

# plotting the points
plt.plot(x_left, y_left, marker='o')

# plotting the points
plt.plot(x_right, y_right, marker='o', color='#1f77b4')

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

plt.grid(linestyle = '--', linewidth = 0.5)

#axis limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)

xorigem_ver = [0, 0]
yorigem_ver = [-6, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-6, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

# function to show the plot
plt.show()
```

Podemos concluir de todo o anterior que usar os limites pode ser uma ferramenta muito útil, e sobre todo, rápida, para averiguar a onde tende uma função qualquer. 

Finalmente, podemos aplicar rapidamente os limites a nossas funções de exemplo para ver qual é sua tendência. No caso da função $f(x)=2x+1$, ela não tem descontinuidades e, portanto, seu domínio será toda a reta real. Dito isso, só resta saber para onde tende quando os valores de $x$ são positiva e negativamente grandes:
> $\lim_{x\to \infty} f(x) = \lim_{x\to \infty} 2x+1 = 2(\infty)+1 = \infty$

> $\lim_{x\to -\infty} f(x) = \lim_{x\to -\infty} 2x+1 = 2(-\infty)+1 = -\infty$

Ou seja, a função cresce infinitivamente quando $x$ aumenta positivamente e decresce infinitivamente quando vamos para o lado oposto para $x$ negativos.

Respeito à função logaritmo natural, acontece algo similar para valores de $x$ grandes e positivos:
> $\lim_{x\to \infty} f(x) = \lim_{x\to \infty} ln(x) = ln(\infty) = \infty$

Em quanto para os valores negativos, já que a função não está definida para os mesmos, podemos estudar o que acontece quando nos aproximamos a zero vindo da direita:
> $\lim_{x\to 0^{+}} f(x) = \lim_{x\to 0^{+}} ln(x) = ln(0^{+}) = -\infty$

Concluímos, após esse calculo de limites, que os gráficos destas funções, levantados nas seções anteriores, poderiam ser desenhados com certeza para o domínio todo dessas funções.

## 6. Inclinação de uma função linear
A função linear está definida por dois parâmetros muito importantes porque apenas com eles podemos levantar a função por completo, mas também porque particularmente um deles será transcendental para entender mais sobre funções. Toda função linear tem esta forma:
> $f(x) = ax+b$

onde o parâmetro $a$ é o chamado **coeficiente angular** ou **inclinação** e o parâmetro $b$ é o chamado **coeficiente linear**, o qual será o ponto onde a função corta a ordenadas. 

Como temos visto nas seções anteriores, quando dispomos analiticamente da função, estudando suas características tal e como foi ensinado, podemos averiguar toda a informação possível. Mas cabe a seguinte pergunta: e se por acaso não temos tal função, mas só apenas alguns pontos da mesma? Isso é de fato o mais habitual em Ciência. Por isso, um passo muito importante para ter a função seria poder reconstruir a mesma a partir de seus pontos ou de alguma informação relacionada.

No caso da função linear, podemos facilmente averiguar como é a função a partir de alguns pontos do gráfico. De fato, com apenas dois pares ordenados podemos averiguar tanto a inclinação da reta da função linear (o coeficiente angular) quanto o ponto de corte em ordenadas (o coeficiente angular).

A inclinação de uma reta se define comparando um intervalo vertical, o intervalo dos valores da função, com um intervalo horizontal, o invervalo dos valores de $x$:
> $i=\frac{\Delta y}{\Delta x}=\frac{\Delta f(x)}{\Delta x}=\frac{f(x_2) - f(x_1)}{x_2 - x_1}$

```python

# importing the required module
from turtle import color
import matplotlib.pyplot as plt

# x axis values
x = [-2,-1,0,1,2,3,4]
# corresponding y axis values
y = [-3,-1,1,3,5,7,9]

# plotting the points
plt.plot(x, y, marker='o')

# x1 values
x1 = [1, 1]
y1 = [0, 3]

# plotting the x1, y1 points
plt.plot(x1, y1, marker='o', color='grey', linestyle="--")

# x2 values
x2 = [3, 3]
y2 = [0, 7]

# plotting the x1, y1 points
plt.plot(x2, y2, marker='o', color='grey', linestyle="--")

# Delta_x values
Delta_x_hor = [1, 3]
Delta_x_ver = [3, 3]

# plotting the x1, y1 points
plt.plot(Delta_x_hor, Delta_x_ver, marker='o', color='black', linestyle="--")

# Delta_fx values
Delta_fx_hor = [3, 3]
Delta_fx_ver = [3, 7]

# plotting the x1, y1 points
plt.plot(Delta_fx_hor, Delta_fx_ver, marker='o', color='black', linestyle="--")


# fx1 values
fx1 = [0, 1]
fy1 = [3, 3]

# plotting the fx1, fy1 points
plt.plot(fx1, fy1, marker='o', color='grey', linestyle="--")

# fx2 values
fx2 = [0, 3]
fy2 = [7, 7]

# plotting the fx2, fy2 points
plt.plot(fx2, fy2, marker='o', color='grey', linestyle="--")

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-2, 4)
plt.ylim(-2, 8)

plt.grid(linestyle = '--', linewidth = 0.5)

xorigem_ver = [0, 0]
yorigem_ver = [-2, 8]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-2, 8]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

label_x1 = 'x$_1$'

plt.text(x1[0], y1[0], label_x1, ha = 'center', va = 'top', size = 12, color = 'grey')

label_x2 = 'x$_2$'

plt.text(x2[0], y2[0], label_x2, ha = 'center', va = 'top', size = 12, color = 'grey')

label_x1 = 'f(x$_1)$'

plt.text(fx1[0], fy1[0], label_x1, ha = 'right', va = 'top', size = 12, color = 'grey')

label_x1 = 'f(x$_2)$'

plt.text(fx2[0], fy2[0], label_x1, ha = 'right', va = 'top', size = 12, color = 'grey')

plt.subplot(111).annotate('$\Delta x$', xy=(3, 3), xytext=(2, 2.5))
plt.subplot(111).annotate('$\Delta f(x)$', xy=(3, 3), xytext=(3, 5))
plt.subplot(111).annotate('$\Theta$', xy=(3, 3), xytext=(1.5, 3.5))

# function to show the plot
plt.show()
```
