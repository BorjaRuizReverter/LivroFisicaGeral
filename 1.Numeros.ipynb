## 1. Introdução
Os números são a base da Ciência. O espetacular desenvolvimento da cultura ocidental durante os últimos séculos foi, de fato, graças aos números. Desde os movimentos do mouse/touchpad que agora estamos fazendo no nosso computador até os scroll que fazemos na tela do nosso celular com o dedo, tudo é convertido em números através de funções de transferência que transformam energia elétrica ou lumínica em números. Como se o filme de Matrix se tratasse, tudo ao nosso redor pode ser convertido em números.

Porém, ao contrário do que indica este mesmo filme, o fato de que possamos realizar esta conversão entre mundo real e mundo matemático não quer dizer que este último realmente exista. A matemática é uma construção mental, desenvolvida por nosso cérebro de maneira natural por meio da lógica, e que é usada para desenvolver a Ciência através da própria lógica e através da experiência.

## 2. Números naturais
Na vida real é muito habitual ter que **contar** coisas. Estas coisas se chamam **elementos**, os quais formam parte da estrutura mais simples que existe em Matemáticas: os **conjuntos**. Os conjuntos são associações de elementos sob uma determinada propriedade. Por exemplo, podemos definir $D$ como o conjunto dos dias da semana:
$$D = \{segunda, ter\textit{\c{c}}a, quarta, quinta, sexta, sábado, domingo \}$$

ou, $M$ como o conjunto de coisas que um estudante, num determinado momento, têm acima da mesa:
$$M = \{computador, mouse, cafe \}$$

ou, $M_{0}$ como o conjunto de coisas que um estudante, num determinado momento, têm acima da mesa, após ter esvaziado a mesma:
$$M_{0} = \{ \}$$

Observe-se que, o conjunto $D$ dos dias da semana, possui 7 elementos, o conjunto $M$ possui 3 e o conjunto $M_{0}$ possui 0 elementos. Efetivamente, dados estes conjuntos de exemplo, ou qualquer outro tipo de conjunto, a operação mais elementar que existe é a operação de contar. Precisamente para isso foi que surgiram os chamados **números naturais**, o qual está formado pelos seguientes elementos:

$$\mathbb{N}_0 = \{0, 1, 2, 3, 4, ... \}$$

Os números naturais são, por sua vez, um conjunto, pois ele está formado pelos elementos, os números, que se usam para contar elementos de qualquer conjunto. Porém, ao contrário dos conjuntos anteriores, ele possui infinitos elementos. **O infinito**, representado por $\infty$ não é um número, mas uma simbologia para denotar que um conjunto, neste caso o conjunto dos números naturais, não tem fim. Realmente, é algo mais do que isso. De fato, o infinito á uma maneira de materializar essa ideia em um símbolo. O uso do infinito pode parecer um tanto estranho, mas, como veremos futuramente, sua materialização resultou em uma das ideias mais brilhantes da Matemática, pois, graças a ela, foi desenvolvido o cálculo integral e diferencial.

Esta maneira de apresentar os números naturais provem da chamada Teoria de Conjuntos desenvolvida principalmente por Georg Cantor. Entretanto, existe uma maneira alternativa de introduzir os números naturais através de certas regras. Isso de fato foi a sugestão de Giuseppe Peano que, através de 5 simples axiomas, deu à luz elegantemente aos números naturais. Estes axiomas se baseiam principalmente em que o número 1 é primeiro número natural e o resto são seus sucessores. Deste ponto de vista, o número 1 seria o elemento gerador, pois o resto dos números naturais se constroem a partir deste. Por exemplo, a letra com a qual representamos o número 2 consiste em uma simplificação de compor o 1 com seu sucessor. Já a letra com a qual representamos o número 3 consiste numa simplificação para denotar a composição do 1 com seu sucessor com seu sucessor. E assim em diante com todas as letras que usamos para representar os números naturais.

Observe-se que, de maneira natural, a operação de sucessão dos números naturais segundo Peano, é quem define a operação da **soma**. De fato, revisando o dito anteriormente, o sucessor do número 1, que define o número 2, se consegue através da operação 1+1; o sucessor do sucessor do número 1 se consegue através da operação 1+1+1; e assim em diante.

Eventualmente, o conjunto dos números naturais segundo Peano seria este:

$$\mathbb{N}_1 = \{1, 2, 3, 4, ... \}$$

onde o zero não está incluido, já que, segundo os axiomas de Peano, o número 1 é o primeiro natural, e este não é o sucessor de ninguém.

Não existe, porém, um consenso sobre qual das definições para gerar os números naturais é a certa. A corrente logicista e da teoria da computação aderem à definição dos números naturais segundo Cantor enquanto a corrente axiomática opta pela abordagem de Peano.

Em qualquer caso, exceto a exclusão do zero, o detalhe mais significativo da introdução dos números naturais segundo Peano é o fato de que os números estão estreitamente ligados às possíveis operações matemáticas que podem ser feitas com eles. Efetivamente, a introdução dos números naturais segundo Peano é possível graças à soma (+), pois a soma é a maneira de encontrar sucessores respeitando os axiomas, isto é, sem sair do conjunto dos números naturais ou sem forcar esse conjunto em outros conjuntos filhos. Operações binárias como 2+3 são, de fato, perfeitamente possíveis, mas não operações de divisão ou subtração como 2/3 ou 2-3, pois estas criam números que não são naturais.

A operação de multiplicação ainda é permitida, pois qualquer número natural vezes outro gera um número que também é natural. Como veremos proximamente, a combinação de certos conjuntos de números com certas operações gera novos conjuntos e ainda estruturas algebraicas cada vez mais complexas que os matemáticos denominam como grupos, anéis, módulos, álgebras, corpos, espaços vetoriais e assim em diante.

O importante, entretanto, não é tanto conhecer em profundidade todas e cada uma dessas estruturas se não resgatar apenas aquilas necessárias para entender a Física e principalmente as contas que emergem das operações com as equações físicas. Para isto devemos nos focar no que foi definido até agora: o conjunto dos números naturais e as operação que podem ser feitas entre eles: a soma e a multiplicação.

A geração dos números naturais através da operação de somar, segundo os axiomas de Peano, transmite a sensação de que o conjunto dos números naturais é algo monótono e enfastiante. Entretanto, tal conjunto possui uma interessante estrutura, a qual não é percebível a priori, quando vista através da operação de multiplicação.

Neste ponto resulta interessante entender a diferença entre uma e a outra. Uma operação como 2+3 realmente consiste em operar (1+1+1)+(1+1) e, portanto, deste ponto de vista, somar não é mais que acrescentar números 1 ao número 1. Já uma operação de multiplicar como 2 $\cdot$ 3 tem a interpretação de somar o número três duas vezes (3+3). Ou seja, de maneira geral, se somar equivale a descrever quantidades de coisas (ex. 7 dias da semana), multiplicar equivale a descrever quantidades de números (ex. 2 vezes 3). A operação de multiplicar é algo mais inerente aos números naturais, pois esta consiste em, digamos, usar números para operar números.

Podemos, precisamente, usar a multiplicação para enxergar essa interessante estrutura dos números naturais. Podemos usar para isso, por exemplo, grãos de arroz, separarando uma determinada pilha de grãos em varias outras pilhas menores iguais desde que sejam diferentes da unidade. Por exemplo, o número 2 só pode ser separado em duas pilhas unidade (2 pilhas de 1 grão cada, ou seja, 2 $\cdot$ 1) e o número 3 em três pilhas unidade (3 pilhas de 1 grão cada, ou seja, 3 $\cdot$ 1), então nenhum deles não satisfazem a condição supracitada. O número 4 pode sim ser separado em duas pilhas menores diferentes de 1 (2 pilhas de 2 grãos cada, isto é, 2 $\cdot$ 2). Já o número 5 só pode ser separado em pilhas de um (5 pilhas de 1 grão cada, aliás, 5 $\cdot$ 1). O número 6 pode sim ser separado segundo essa condição (2 pilhas de 3 grãos cada, ou seja, 2 $\cdot$ 3), etc...

De fato, uma maneira alternativa de construir o conjunto dos números naturais é precisamente através dos números primos, pois todo número natural ou é 1, ou é um número primo, ou é um número composto por números primos. Isto é precisamente o que afirma o chamado **Teorema Fundamental da Aritmética**.

## 3. Números inteiros
Como acabamos de ver, podemos construir o conjunto dos números naturais contando coisas ou somando coisas, o qual, realmente, equivale a repetir quantas vezes quisermos o número 1. Sendo assim, o conjunto dos números naturais forma uma estrutura coerente frente a operação de soma, pois está equivale a compor o número 1 com ele mesmo as vezes necesárias, e também é coerente com a operação de multiplicação, já que esta equivale a somar não números se não quantidades de números.

Existe, porém, outra operação conhecida como **subtração**, que cria certas incongruências quando usada com números naturais. Por exemplo, é claro que 3 menos 2 é 1, mas qual seria o resultado de 2 menos 3? A operação de subtração da raiz aos **números negativos**, os quais estão fora do conjunto dos números naturais. A operação de subtração da a luz a um novo conjunto mais abrangente, que possibilite realizar subtrações que estejam em todo momento definidas. Este conjunto é chamado conjunto dos **números inteiros**, denotados pela letra $\mathbb{Z}$, e composto pelos seguintes números:

$$\mathbb{Z} = {-\infty, ..., -4, -3, -2, -1, 0, 1, 2, 3, 4, ..., \infty }$$

Resulta interessante reparar como, a partir da operação de subtração e a partir dos números naturais, emerge o **o número zero**. Sendo mais específicos, o número zero é criado subtraindo um número qualquer com ele mesmo. Por exemplo,
$$3 - 3 = 0$$
Operação que também pode ser entendida como:
$$3 + (- 3) = 0$$
