### 1. Introdução
A Física é uma ciência experimental. Toda ciência experimental consiste na observação. Os físicos observam os entes da Natureza ou os fenômenos protagonizados por estes para, a partir deles construir uma explicação teorica baseada na lógica, isto é, na matemática. A matemática é portanto a linguagem da física, e por ende, a linguagem da Natureza.

A Física não é e nunca será uma ciência completa ou acabada. Os entes da Natureza atuam dentro de escala particular de tempo e espaço. Porem, se aprofundamos o estudo destes entes diminuindo dita escala, observa-se que o mesmo ente se dilui em outros entes, dos quais o primeiro se compõe, em um processo que pode ser repetido infinitamente*. Analogamente, ao estudar este ente em escalas maiores observa-se que o mesmo forma parte de uma estrutura maior, a qual também não tem por que ter limite. A estrutura da Natureza, portanto, são como as camadas de uma cebola: partes (ou entes) que se compõem de outras partes, que se compõem de outras partes, que se compõem de outras partes...

Em qualquer caso, no nivel de estudo desta disciplina de física básica, isso não deve nos preocupar, pois os entes da escala de tempo e espacio que usaremos será aquela particular da nossa vida cotidiana, asociada, como veremos a um tipo especial de sistema de unidades.

### 2. Grandezas físicas

Deixando de lado a estrutura da Natureza (mas tenendo-a em conta caso aprofundemos no estudo da Física), uma pergunta que cabe é: se a Física é uma ciência baseada na observação da Natureza, o que observamos? E como observamos? A observação consiste em comparar o objeto observado com outros objetos conhecidos através de suas propriedades qualitativas (QUALIS). Por exemplo, se o objeto desconhecido que observamos move-se como algo familiar a um cachorro, come como um cachorro e late como um cachorro então concluimos que esse objeto é com alta probabilidade um cachorro, e não, digamos, um gato. E chegamos a essa conclusão simplesmente porque o ente, a priori desconhecido, possui as qualidades de outro conhecido: um cachorro. Esta abordagem parace absurda, mas é o que nosso cérebro faz para reconher objetos ou fenômenos desconhecidos. Os algarismos de AI (Artificial Intelligence), tão usados hoje em dia, estão de fato baseados nessa ideia: comparar as qualidades de objetos novos, recolhidos por uma câmara, para depois etiqueta-los e tomar a decisão de se o objeto é conhecido, graças ao treinamento do algoritmo, ou não.

O problema na hora de identificar as qualidades dos objetos é que estas qualidades podem ser de dois tipos: objetivas ou subjetivas. As qualidades subjetivas dependem do sujeito que identifica essa qualidade, enquanto que as qualidades objetivas não. Por exemplo, é claro que a altura de um cachorro é uma qualidade objetiva, pois qualquer individio, seja ele brasileiro, chinese ou um morador de outro planeta, vai concordar na mesma observação da altura. Propriedades como a altura, a massa, a velocidade, a resistência elétrica, o volume ou a temperatura do cachorro são claramente qualidades objetivas e estas as chamamos grandezas físicas. Por outro lado, qualidades como o cheiro, a alegria ou o carinho são subjetivas, pois estas são avaliadas diferentemente por cada individuo.

Esta separação entre qualidades objetivas e subjetivas porem é fundamental e provoca um claro cisma entre o que é conhecimento universal e verdadeiro e o que não é. As qualidades objetivas são os tijolos da Ciência. As qualidades subjetivas são os tijolos de outras áreas pseudo-cientificas, a pesar de que estas se orgulhem de se chamar também de Ciências, quando não são.

### 3. Padrões e incertezas inerentes

Para transformar a observação da Natureza em informação objetiva repoduzível por outra pessoa devemos medir, e medir é comparar. Ou seja, ao medir estaremos compararando a qualidade objetiva de nosso objeto de estudo com a mesma qualidade de outro objeto padrão conhecido. Por exemplo, se quiser reportar para outra pessoa no Brasil qual é altura do cachorro em questão posso usar notas de 10 reais colocadas em horizontal e contar (QUANTIS) qual seria a altura do cachorro em notas deitadas. Desta maneira, posso declarar, por exemplo, que a altura do chachorro problema é de algo mais de 5 notas de 10 reais deitadas, mas não chega a 6.

Mas foram 5 notas ou 6 notas? A altura está mais perto de 5 ou de 6 notas? Podemos ser mais precisos? Sim e não. Poderiamos por exemplo usar outro objeto padrão (ou padrão diretamente) como por exemplo um dedo. Assim poderiamos voltar a medir a altura do chachorro usando o padrão polegadas e afirmar que ele mede algo entre 25 e 26 polegadas. Mas foram 25 ou 26 polegadas? Então, vemos agora que, mesmo mudando o padrão, sempre vamos ter essa incerteza, a qual por outro lado é inerente ao processo de medida. Medir implica errar, mas esse erro pode ser estudado dentro do campo da física experimental, a qual fica fora de nosso escopo. No que repeita a este texto de Física Geral, diremos por exemplo que altura de um objeto é 25 polegadas, ou 25.4 ou 25.43, ou 25.437 a pesar de que em todos os casos, o último algarismo seja sempre incerto. Para contornar esse problema inerente às medidas, alguns livros de Física reportam os números como 25.0 polegadas, ou 25.40, ou 25.430, ou 25.4370, querendo dizer que todos os algarismos diante do zero são certos e o único incerto é o zero.

Voltando aos padrões usados para medir, notas de 10 reais deitadas e as polegadas, observe-se como ambas apresentam problemas. O primeiro só pode ser usado no Brasil enquanto que o segundo é um padrão que depende do observador, pois cada pessoa tem um dedo diferente. Ao usar estes padrões voltamos a cair na subjetividade inicial e isso representa um problema. Então, o lógico é usar padrões que sejam universais e invariáveis, e que também sejam reproduzíveis por qualquer um que deseje usá-lo.

Observe-se que nessa afirmação há implicita duas informações: o QUANTIS, indicando explicitamete pelo número de notas de 10 reais, e o QUALIS, indicado implicitamente pelo objeto padrão que foi usado para medir, neste caso, de novo, a nota de 10 reais deitada. Então, note-se que ao dizer qual é objeto padrão (ou padrão diretamente) que foi usado para medir, já implicitamente estou informando a grandeza física usada:  altura (comprimento).

Dito de outro modo, as propriedades qualitativas objetivas podem ser medidas por exemplo usando algum outro objeto padrão (polegadas, pés, notas de 10 reais colocadas em vertical,...) e essa informação, junto com a informação do objeto padrão que foi usado para medir, pode ser repasada para outra pessoa, a qual podría reproduzir subsequentemente qual seria a altura do cachorro em questão. Por exemplo, poderiamos medir que a altura de um cachorro A é de 3 notas de dez reais deitadas e que a altura de um chachorro B é de 5 notas de dez reais deitadas. Dessa maneira, mediante a medição, poderiamos comparar os valores de uma determinada grandeza física e concluir objetivamente que o cachorro A é mais alto que o cachorro B.

Concluimos pois, que para descrever a Natureza devemos usar um dupla de informações dada pelo QUALIS (uma qualidade objetiva) e pelo QUANTIS (as vezes que essa qualidade objetiva se repete quando comparada com um padrão de medida). Desta maneira, em Física todos os números serão expresos, por exemplo, da seguinte forma:
> $$25\cdot pol$$

onde $pol$ faz referencia ao QUALIS que estamos medendo (neste caso um comprimento), junto com o padrão que estamos usando para medir esse comprimento (neste caso polegadas) e o número $25$ faz referencia ao número de vezes que o padrão usado para medir comprimento se repete.

### 4. Sistemas de unidades
Como vimos anteriormente, podemos usar qualquer objeto padrão para medir, mas existem uns padrões melhores que outros. Para diminuir o grau de incerteza na medição e aumentar o rigor convem usar padrões universáis, invaríaveis e facilmente reproduzíveis. Chegados neste ponto, a questão é: quem estabelece o melhor padrão? Por que um padrão específico e não outro? Com qual criterio?

Não existe um criterio absoluto e objetivo para estabelecer os melhores padrões. Por isso, a única saída possível é chegar a um consenso através de uma convenção. Um dos convenios mais conhecidos é a Conférence générale des poids et mesures, encarregada de avaliar e gerir o conjunto de padrões mais usado: o **Sistema Métrico Decimal**, o que desembocou no chamado **Sistema Internacional de Unidades**.

O Sistema Internacional de Unidades (**S.I.** em diante) está atualmente formado por 7 unidades fundamentais, das quais se derivam o resto das unidades conhecidas, chamadas por isso unidades derivadas.

### 5. Notação Científica
Como explicamos anteriormente, o objetivo da Física é descrever a Natureza a partir da observação. Porém, nessa observação podemos nos deparar com que os objetos que observarmos possuem tamanhos enormemente grandes (como, por exemplo, o comprimento da Via Láctea, de valor 9 500 000 000 000 000 000 km) ou increívelmente pequenos (como o ráio atômico do Hidrogênio de valor 0,000 000 000 012 m), ou agem dentro de intervalos temporais enormes (como a rotação de um sistema solar ao redor do centro da galaxia de valor 230 000 000 anos) ou surpreendentemente pequenos (como as vibrações moleculares, de valor ao redor de 0,000 000 000 000 01 segundos).

Lidar com esse tipo de números, às vezes tão grandes e outras tão pequenos não é nada confortável. Já que estes números tipicamente contem um número enorme de zeros, o ideal seria usar uma notação onde esses zeros fossem sintetizados. Aqui é onde entra a chamada **Notação Científica**, a qual consiste em simplificar os números grandes ou pequenos tirando proveito da base decimal, ou base 10.

Para transformar um número grande ou pequeno em outro em base 10. Devemos proceder da seguinte maneira. Escolhamos inicialmente um número muito grande como o anterior dado do comprimento da Via Láctea:
> $9 500 000 000 000 000 000 \cdot km$

Em primeiro lugar, criamos uma base 10:
> $9 500 000 000 000 000 000 \cdot km = 9 500 000 000 000 000 000 \cdot 1 \cdot km = 9 500 000 000 000 000 000 \cdot 10^0 \cdot km$

E em segundo lugar, mudamos a vírgula para esquerda, removendo zeros, ao tempo que, para compensar, avançamos com o exponente da base 10 para direita:
> $9 500 000 000 000 000 000. \cdot 10^0 \cdot km = 9 5. \cdot 10^{17} \cdot km = 9 5 \cdot 10^{17} \cdot km$

E desta maneira, teremos transformado um número muito grande em notação científica.

Se, ao contrário, o número for muito pequeno procedemos de maneira análoga, mas invirtindo os movimentos: vírgula para direita e exponente da base 10 para esquerda. Usemos o número do raio atômico supracitado para conferir:
> $0.000 000 000 012 \cdot m =  0.000 000 000 012 \cdot 1 \cdot m = 0.000 000 000 012 \cdot 10^0 \cdot m = 12 \cdot 10^{-12} \cdot m$

### 6. Álgebra de Notação Científica
Ao longo desta disciplina de Física Geral vamos nos deparar com equações envolvendo grandezas físicas. Para resolver essas equações teremos que substituir os valores de tais grandezas físicas, tipicamente números em formato de notação científica, e operar esses números. Podemos dividir essas operações em básicamente três: (1) a multiplicação e divisão, (2) a exponenciação e (3) a soma e subtração. Vejamos tais operações em detalhe através de exemplos práticos.

#### 6.1 Mutiplicação e divisão de números em notação científica
Suponhamos que, por algum motivo, precisemos, a partir do valor do comprimento dos lados, calcular uma área. A modo de exemplo, suponhamos que um lateral mede 2 10³ m e o outro mede 3 10² m. Sendo assim, o cálculo da área *A* seria:
> $A = comprimento \cdot comprimento = (2 \cdot 10^3 \cdot m) \cdot (3 \cdot 10^2 \cdot m) = 2 \cdot 10^3 \cdot m \cdot 3 \cdot 10^2 \cdot m = 2 \cdot 3 \cdot 10^3 \cdot 10^2 \cdot m \cdot m = 6 \cdot 10^5 \cdot m^2$

Observe que, nessa operação, houve três detalhes importantes. Primeiro, é usar sempre os parénteses na hora substituir um valor na equação. Segundo, é usar a propriedade comutativa, pois a ordem dos fatores (fatores são números, ou unidades, que multiplicam ou dividem) não altera o produto. Isso é fundamental para juntar entes similares para depois operar entre eles: juntar números com números, bases com bases e unidades com unidades, mantendo assim a estrutura de um número em notação científica. E terceiro, lembrar da propriedade matemática na qual os exponentes somam quando a base é a mesma.

Procedamos agora com a divisão. Suponhamos que em algum problema sobre Dinâmica precisemos, dados os valores de uma única força F aplicada sobre um corpo e a massa m do mesmo, averiguar a aceleração que o corpo experimentaria. A modo de exemplo, suponhamos que dita força vale 2 10³ N, onde N significa Newtons, ou seja, a unidade padrão da grandeza física força, como indica a TABELA 2, e que a massa vale 3 10² kg. Sendo assim, o valor da aceleração seria:
> $a = \frac{F}{m} = \frac{(2 \cdot 10^3 \cdot N)}{(3 \cdot 10^2 \cdot kg)} = \frac{2 \cdot 10^3 \cdot kg \cdot \frac{m}{s²}}{3 \cdot 10^2 \cdot kg} = \frac{2}{3} \cdot \frac{10^3}{10^2} \cdot \frac{m}{s^2} = \frac{2}{3} \cdot 10^3 \cdot 10^{-2} \cdot \frac{m}{s^2}$

Vemos que aqui o detalhe principal foi que uma operação de divisão, no que respeita às bases de 10, se convertiu numa operação de multiplicação, na qual os exponentes de bases iguais somam. Desta maneira, percebemos que dividir equivale a multiplicar pelo inverso do número que está no denominador. E o inverso de um número, ou de uma unidade pois se comporta igual, construi-se mudando o sinal do seu exponente. Alias, tomando como x um número qualquer, o inverso de x será:
> $\frac{1}{x} = x^{-1}$

Insistimos em que esta mesma operação de tomar o inverso, feito com números, pode ser também feita para as unidades. Assim, se em alguma operação tiver que calcular uma frequência (f) a partir de um comprimento de onda ($\lambda$) de, suponhamos, 2 10<sup>-2 </sup>m, o resultado seria:
> $f = \frac{1}{\lambda} = \frac{1}{2 \cdot 10^{-2}m} = \frac{1}{2}\frac{1}{10^{-2}}\frac{1}{m} = 0,5 \cdot 10^2 \cdot m^{-1} = 5 \cdot 10^1 \cdot m^{-1} = 50 Hz$

já que Hz, ou Hertz, o nome que se da às unidades m<sup>-1</sup>, ou seja, às unidades de frequência.

#### 6.2 Exponenciação de números em notação científica
No mesmo capítulo de Dinâmica, teremos que calcular propriedades da partícula, como sua energia cinética (K), a qual depende da massa (m) da partícula e de sua velocidade (v) ao quadrado. Supondo que tanto m quanto v são dados, de valores, por exemplo, 2 10³ kg e 3 10² m/s² respectivamente, o resultado da energia cinética seria:
> $K = \frac{1}{2} m v^2 = \frac{1}{2} (2 \cdot 10^3 \cdot kg) \cdot (3 \cdot 10^2 \frac{m}{s})^2 = \frac{1}{2} (2 \cdot 10^3 \cdot kg) \cdot (3^{1 \cdot 2} \cdot 10^{2 \cdot 2} \frac{m^{1 \cdot 2}}{s^{1 \cdot 2}}) = \frac{1}{2} (2 \cdot 10^3 \cdot kg) \cdot (3^2 \cdot 10^4 \frac{m^2}{s^2}) = \frac{1}{2} \cdot 2 \cdot 3^{2} \cdot 10^3 \cdot 10^4 \frac{kg \cdot m^{2}}{s^{2}} = 9 \cdot 10^7 \frac{kg \cdot m^{2}}{s^{2}} = 9 \cdot 10^7 J$

, sendo os Joules (J) a unidade associada a grandeza física energia. Observe-se em qualquer caso o detalhe principal nesta operação: ao elevar um número em notação científica a qualquer exponente, os exponentes de todos os fatores deve ser multiplicado pelo exponente externo.

#### 6.3 Soma e subtração de números em notação científica
A última operação lícita entre números em notação científica é a soma/subtração. Já que um número em notação científica está composto por, no máximo, 3 fatores (o número mesmo, a base 10 e a unidade) a maneira de proceder numa soma/subtração procurar ou, se for possível criar, fatores em comum aos dois somandos e deixar eles em evidência. Vejamos com um exemplo. A equação fundamental da hidrostática relaciona a pressão (p) a uma profundidade específica, com o valor dessa profundidade (h), com a densidade do fluido ($\rho$), com a aceleração da gravidade (g $\approx$ 10 m\s²) e com a chamada pressão de referência (p<sub>0</sub>). Supondo os seguintes valores: h = 1 m (de profundidade), $\rho$ = 10³ kg/m³ (a densidade da agua), e p<sub>0</sub> = 10<sup>5</sup>Pa (a pressão atmosférica no nível do mar), o resultado dessa conta seria:
> $p = p_0 + \rho \cdot g \cdot h = 10^5 \cdot Pa + 10^3 \frac{kg}{m^3} \cdot 10 \frac{m}{s^2} \cdot 1 \cdot m = 10^5 \cdot Pa + 10^4 \frac{kg}{m s^2} = 10^5 \cdot Pa + 10^4 \cdot Pa = (10^5 + 10^4) Pa = (10 \cdot 10^4 + 10^4) Pa = (10 + 1) 10^4 \cdot Pa = 11 \cdot 10^4 \cdot Pa$

Observe-se portanto, como foi aplicado o explicado anteriormente: na hora de somar devemos ir procurando, ou criando, fatores em comum e deixando-os em evidência.

Finalmente, para efetuar uma subtração de números em notação científica devemos lembrar do elemento oposto: seja qualquer número, ou unidade, y, seu oposto será -y. Dito isso, devemos visualizar que subtração não é mais que somar um número com seu oposto. Ou seja,
> $x - y = x + (-y)$

e, portanto, a operação de subtração entre dois números é análoga à soma. E sendo assim, todas as contas explicadas anteriormente seriam as mesmas com a única mudança de substituir um símbolo de somar pelo de subtrair e ter que efetuar dita subtração final. Podemos usar um exemplo envolvendo a energia total, a qual, como veremos, será a soma da energia cinética com a energia potencial. Suponhamos que a energia cinética (K) tenha um valor de 2 10³ J enquanto que a energia potencial (U) tenha um valor de -3 10² J. Desta manera, a energia total (E) será:
> $E = K + U = (2 \cdot 10^3 \cdot J) + (-3 \cdot 10^2 \cdot J) = (2 \cdot 10^3 -3 \cdot 10^2)J = (2 \cdot 10 \cdot 10^2 -3 \cdot 10^2)J = (2 \cdot 10 -3)10^2 \cdot J = 17 \cdot 10^2 \cdot J$

### 8. Transformação de unidades
Dado que, como vimos anteriormente, dispomos de varios sistemas de unidades e, por sua vez, dentro de cada sistema, encontramos uma serie de prefixos, ou seja, subunidades de cada unidade, é imprescindível saber como **transformar unidades** de um número. Aqui é onde entram os chamados **fatores de conversão**, que será a técnica que usaremos para transfomar unidades, a qual está baseada no uso da notação científica e, especificamente, nas bases de 10.

A técnica dos fatores de conversão consiste em procurar fatores que relacionem unidades diferentes, as quais queremos transformar. Por exemplo, se quisermos efetuar uma transformar entre unidades de comprimento, por ejemplo, entre km e m, precisamos de uma relação simples entre essas unidades. A relação mais simples e conhecida é:
> $1 \cdot km = 10^3 \cdot m$

Agora, tal equivalência entre as unidades deve ser convertida em um fator de conversão, o qual é o que usaremos para efetivar a transformação. Já que o número 1 é o neutro da multiplicação, podemos então proceder assim:
> $1 \cdot 1 \cdot km = 10^3 \cdot m$

e isolando o número 1 que foi colocado à esquerda,
> $1 = \frac{1 \cdot km}{10^3 \cdot m}$

A parte que restou à direita do igual se chama **fator de conversão direto** e, como vemos, ele vale a unidade! Isso será muito importante porque quer dizer que, como o fator de conversão vale a 1, podemos insertar ele nas contas e transformar.

Da mesma maneira que de uma relação entre unidades foi extraido um fator de conversão direto, pode-se também extrair o chamado **fator de conversão inverso**. Para isso, desta vez, insertamos o número 1 à direita do igual:
> $1 \cdot km = 10^3 \cdot m \cdot 1$

e isolamos ele,
> $1 = \frac{10^3 \cdot m}{1 \cdot km}$

Finalmente, agora sim podemos nos focar mesmo na transformação. Caso quisermos transformar 1 km para m devemos usar o fator de conversão inverso, pois só este é quem faz cortar as unidades de km:
> $ 1 \cdot km = 1 \cdot km \cdot 1 = 1 \cdot km (\frac{10^3 \cdot m}{1 \cdot km}) = 1 \cdot km \frac{10^3 \cdot m}{1 \cdot km} = \frac{10^3 \cdot m \cdot km}{km} = 10^3 \cdot m$

Observe como o fator de conversão foi insertado nas contas sem alterar as mesmas já que ele vale a unidade. Observe também que em caso de efetuar a transformação inversa, ou seja, transformar 1 m para km, devemos usar o fator de conversão inverso, pois ele que faz apagar as unidades de m:
> $ 1 \cdot m = 1 \cdot m \cdot 1 = 1 \cdot m (\frac{1 \cdot km}{10^3 \cdot m}) = 1 \cdot m \frac{1 \cdot km}{10^3 \cdot m} = \frac{km \cdot m}{10^3 \cdot m} = 10^{-3} \cdot km$

\
Como vemos, a técnica dos fatores de conversão converte unidades de maneira limpa e elegante, pois não precisamos sair das contas, ao contrário da técnica da regra de três. Entretanto, até agora, todo o anterior poderia ter sido feito usando esta segunda técnica, deixando em dúvida a utilidade da primeira. Na verdade, o verdadeiro potencial da técnica dos fatores de conversão ainda não explicado, pois é ai onde dita técnica se destaca, superando claramente à regra de três. Vejamos alguns exemplo desse potencial.

Uma das grandes vantagens da técnica dos fatores de conversão é que podemos realizar não só uma transformação, se não varias ao mesmo tempo. Vejamos um exemplo disso transformando 1 ano em segundos. Desta vez, vamos insertar diretamente as relações entre as unidades já como fatores de conversão.
> $1 \cdot ano = 1 \cdot ano \cdot 1 \cdot 1 \cdot 1 \cdot 1 \cdot 1 = 1 \cdot ano \cdot (\frac{12 \cdot mês}{1\cdot ano}) (\frac{3 \cdot 10 \cdot dia}{1 \cdot mês}) (\frac{24 \cdot h}{1 \cdot dia}) (\frac{6 \cdot 10 \cdot min}{1 \cdot h}) (\frac{6 \cdot 10 \cdot s}{1 \cdot min})= 12 \cdot 3 \cdot 24 \cdot 6 \cdot 6 \cdot 10^3 \frac{ano \cdot mês \cdot dia \cdot h \cdot dia \cdot min \cdot s}{ano \cdot mês \cdot dia \cdot h \cdot dia \cdot min} = 31104 \cdot 10^3 \cdot s$

Como vemos, a conta toda foi feita elegantemente numa linha, algo que não é possível com a regra de três.

\
Finalmente, a técnica dos fatores de conversão se destaca na hora de realizar transformações nas quais o exponente das unidades é diferente de 1. Vejamos através de um exemplo típico, por exemplo, como transformar de g/cm³ para kg/m³.
> $1 \cdot \frac{g}{cm^3} = 1 \cdot \frac{g}{cm^3} \cdot 1 \cdot 1 = \frac{g}{cm^3} \frac{1 \cdot kg}{10^3 \cdot g} (\frac{10^2 \cdot cm}{1 \cdot m})^3 = \frac{g}{cm^3} \frac{1 \cdot kg}{10^3 \cdot g} (\frac{10^6 \cdot cm^3}{1^3 \cdot m^3}) = 1 \cdot \frac{10^6}{10^3} \frac{g \cdot kg \cdot cm^3}{cm^3 \cdot g \cdot m^3} = 10^3 \frac{kg}{m^3}$

Note-se que, como o fator de conversão vale a unidade, ele pode ser elevado à potência que precisemos. Isso é uma grande vantagem quando na transformação estejam envolvidas unidades elevadas a qualquer potência.
