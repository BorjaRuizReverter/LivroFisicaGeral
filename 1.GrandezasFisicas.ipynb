## 1. Introdução
A Física é uma ciência experimental. Toda ciência experimental consiste na observação. Os físicos observam os entes da Natureza ou os fenômenos protagonizados por estes para, a partir deles construir uma explicação teórica baseada na lógica, isto é, na matemática. A matemática é, portanto, a linguagem da física, e em consequência, a linguagem da Natureza.

A Física não é e nunca será uma ciência completa ou acabada. Os entes da Natureza atuam dentro da escala particular de tempo e espaço. Porem, se aprofundamos no estudo destes, diminuindo dita escala, observa-se que o mesmo ente se dilui em outros menores, dos quais se compõe, em um processo aparentemente sem fim, o qual se especula que poderia na acabar no tempo e na escala de Planck. Analogamente, ao estudar um determinado ente em escalas maiores observa-se que o mesmo forma parte de uma estrutura maior, cujo limite também é desconhecido. A estrutura da Natureza, portanto, são como as camadas de uma cebola: partes que se compõem de outras partes, que se compõem de outras partes, que se compõem de outras partes...

Em qualquer caso, no nível de estudo desta disciplina de Física Geral, isso não será um problema, pois, os entes que estudaremos estarão dentro de uma escala de tempo e espacio, as quais formam parte da nossa vida cotidiana. Isto é, usaremos um sistema de unidades simples e familiar.

## 2. Grandezas físicas

Deixando de lado a estrutura da Natureza (mas tendo-a em conta caso aprofundemos no estudo da Física), uma pergunta que cabe é: se a Física é uma ciência baseada na observação da Natureza, o que realmente observamos? E como é feita essa observação? A observação consiste em comparar o objeto observado, a priori desconhecido, com outros objetos conhecidos através de suas **propriedades qualitativas**. Por exemplo, se o objeto desconhecido que observamos move-se como algo familiar a um cachorro, come como um cachorro e late como um cachorro então concluímos que esse objeto é com alta probabilidade um cachorro, e não, digamos, um gato. E chegamos a essa conclusão simplesmente porque o ente, a priori desconhecido, possui as qualidades de outro conhecido: um cachorro. Esta abordagem parece ridícula, mas é o que realmente nosso cérebro faz para reconhecer objetos ou fenômenos desconhecidos. Os algarismos de AI (Artificial Intelligence), tão usados hoje em dia, estão de fato baseados nessa ideia: comparar as qualidades de objetos novos, recolhidos por uma câmara, para posteriormente etiqueta-los e daí tomar a decisão de se o objeto é conhecido, graças ao treinamento do algoritmo, ou não.

O problema na hora de identificar as qualidades dos objetos é que estas qualidades podem ser de dois tipos: objetivas ou subjetivas. As qualidades subjetivas dependem do sujeito que identifica essa qualidade, enquanto que as qualidades objetivas não. Por exemplo, é claro que a altura de um cachorro é uma qualidade objetiva, pois qualquer individuo, seja ele brasileiro, chines ou um morador de outro planeta, vai concordar na mesma observação da altura. Propriedades como a altura, a massa, a velocidade, a resistência elétrica, o volume ou a temperatura do cachorro são claramente qualidades objetivas e, por tanto, são chamadas de **grandezas físicas**. Por outro lado, qualidades como o cheiro, a alegria ou o carinho são subjetivas, pois estas são avaliadas diferentemente por cada indivíduo.

Esta separação entre qualidades objetivas e subjetivas é fundamental e provoca um claro cisma entre o que é conhecimento universal e verdadeiro e o que não é. As qualidades objetivas são os tijolos da Ciência. As qualidades subjetivas são os tijolos de outras áreas pseudo-cientificas, a pesar de que estas se orgulhem também de se chamarem de Ciências, quando não o são.

## 3. Padrões e incertezas inerentes

Para transformar a observação da Natureza em informação objetiva reproduzível por outra pessoa devemos medir, e **medir consiste em comparar**. Ou seja, ao medir estaremos comparando a qualidade objetiva de nosso objeto de estudo com a mesma qualidade de outro objeto padrão conhecido. Por exemplo, podemos reportar para outra pessoa no Brasil qual é altura do cachorro em questão usando notas de 10 reais colocadas horizontalmente e contar (QUANTIS) qual seria tal altura. Desta maneira, posso declarar, por exemplo, que a altura do cachorro problema é de algo mais de 6 notas de 10 reais deitadas, mas não chega a 7.

Mas foram 6 notas de 10 reais ou foram 7 notas? A altura está mais perto de 6 ou de 7 notas? Podemos ser mais precisos? Sim e não. Poderíamos, por exemplo, usar outro objeto padrão (ou **padrão** diretamente) como, por exemplo, um dedo. Assim poderíamos voltar a medir a altura do cachorro usando o padrão polegadas e afirmar que ele mede algo entre 26 e 27 polegadas. Mas foram 26 ou 27 polegadas? 

A partir deste raciocínio entendemos que, mesmo mudando o padrão, sempre vamos ter essa **incerteza**, a qual, por outro lado, é **inerente ao processo de medida**. Medir implica errar, mas esse erro deve ser estudado dentro do campo da física experimental, o qual está fora de nosso escopo. No que repeita a este texto de Física Geral, diremos, por exemplo, que altura de um objeto é 26 polegadas, ou 26.4 ou 26.43, ou 26.437 a pesar de que em todos os casos, o último algarismo seja sempre incerto. Para contornar esse problema inerente às medidas, alguns livros de Física reportam os números como 26.0 polegadas, ou 26.40, ou 26.430, ou 26.4370, querendo dizer que todos os algarismos diante do zero são certos e o único incerto é o zero.

Voltando aos padrões usados para medir, notas de 10 reais deitadas e polegadas, podemos observar como ambas apresentam problemas. O primeiro padrão infelizmente apenas pode ser usado no Brasil enquanto o segundo é um padrão que depende do observador, pois cada sujeito tem um tamanho de dedo diferente. Ao usar estes padrões, portanto, voltamos a cair na subjetividade inicial e isso representa um problema para construir Ciência. Então, seria mais conveniente usar padrões que sejam universais e invariáveis, e que também sejam reproduzíveis por qualquer um que deseje usá-lo.

Observe-se que nessa afirmação há implícita outras duas informações: o **quantis**, indicando explicitamente pelo número de notas de 10 reais que foram usadas, e o **qualis**, indicado implicitamente pelo objeto padrão que foi usado para medir, neste caso, a nota de 10 reais deitada. Então, note-se que ao dizer qual é o padrão que foi usado para medir, implicitamente já está sendo informada a grandeza física a medir: a altura (comprimento).

Dito de outro modo, as propriedades qualitativas objetivas podem ser medidas, por exemplo, usando qualquer objeto padrão conhecido(polegadas, pés, notas de 10 reais colocadas em horizontal ou em vertical, ...) e essa informação, junto com a informação do objeto desconhecido que está sendo medido, pode ser repassada para outro indivíduo, quem poderia reproduzir o experimento e determinar qual seria a altura do cachorro em questão. Por exemplo, poderíamos medir a altura de um primeiro cachorro, chamemos-o  de cachorro A, suponhamos de 3 notas de dez reais deitadas, e que a altura de um segundo cachorro, chamemos-o de cachorro B, é de 5 notas de dez reais deitadas. Dessa maneira, mediante o processo de medida, podemos comparar grandezas físicas de determinados entes e concluir objetivamente que, neste caso, o cachorro A é mais alto que o cachorro B.

De todo o anterior deduzimos que para descrever a Natureza devemos usar uma **dupla de informações**, as quais estão dadas pelo *qualis* (uma qualidade objetiva) e pelo *quantis* (quantas vezes que essa qualidade objetiva repete o padrão que foi usado para medir). É por isso que em Física todos os números serão expressos, por exemplo, da seguinte forma:

$$26\cdot pol$$

onde $pol$ faz referencia ao qualis que estamos medendo (neste caso um comprimento), o qual confirma também o padrão que estamos usando para medi-lo (neste caso polegadas), e o número $26$ faz referencia ao número de vezes que o padrão usado para medir comprimento se repete.

É importante resaltar aqui um vizio muito comum nos aluno iniciantes em Física, o qual consiste em fornecer apenas o número e não o padrão que foi usado para medir. Observe-se, porem, que isso cria uma ambiguidade enorme. Insistimos em que um número representa quantas vezes o padrão se repete. Portanto, se não colocamos qual padrão, o número por se só não faz sentido.

## 4. Sistemas de unidades
Como vimos anteriormente, podemos usar qualquer objeto padrão para medir, mas existem uns padrões melhores que outros. Para diminuir o grau de incerteza na medição e aumentar o rigor convém usar padrões universais, invariáveis e facilmente reproduzíveis. Chegados neste ponto, a questão é: quem estabelece o melhor padrão? Por que um padrão específico e não outro? Com qual critério?

Não existe um critério absoluto e objetivo para estabelecer os melhores padrões. Por isso, a única saída possível é chegar a um consenso através de uma convenção. Um dos convênios mais conhecidos vem dada pela *Conférence générale des poids et mesures*, encarregada de avaliar e gerir o conjunto de padrões mais usado: o **Sistema Métrico Decimal**, o que desembocou no chamado **Sistema Internacional de Unidades**.

O Sistema Internacional de Unidades (**S.I.** em diante) está atualmente formado por **7 unidades fundamentais**, das quais se derivam o resto das unidades conhecidas, chamadas por isso **unidades derivadas**.

## 5. Notação Científica
Como explicamos anteriormente, o objetivo da Física é descrever a Natureza a partir da observação. Porém, nessa observação podemos nos deparar com que os objetos que observarmos possuem tamanhos enormemente grandes (como, por exemplo, o comprimento da Via-Láctea, de valor 9 500 000 000 000 000 000 km) ou increivelmente pequenos (como o raio atômico do Hidrogênio de valor 0.000 000 000 012 m), ou atuam em intervalos temporais enormes (como a rotação de um sistema solar ao redor do centro da galaxia de valor 230 000 000 anos) ou surpreendentemente pequenos (como as vibrações moleculares, de valor ao redor de 0.000 000 000 000 01 segundos).

Lidar com esse tipo de números, às vezes tão grandes e outras tão pequenos não é confortável. Já que estes números tipicamente contem um número enorme de zeros, o ideal seria usar uma notação onde esses zeros fossem sintetizados. Aqui é onde entra a chamada **Notação Científica**, a qual consiste em simplificar os números grandes ou pequenos tirando proveito da base decimal, ou base 10.

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<p>Para transformar um número grande ou pequeno em outro em base 10. Devemos proceder da seguinte maneira. Escolhamos inicialmente um número muito grande como o anterior dado do comprimento da Via-Láctea:</p>

<p>

$9 500 000 000 000 000 000 \cdot km$

</p>

<p>Em primeiro lugar, criamos uma base 10:</p>

<p>

$9 500 000 000 000 000 000 \cdot km = 9 500 000 000 000 000 000 \cdot 1 \cdot km = 9 500 000 000 000 000 000 \cdot 10^0 \cdot km$

</p>

<p>E em segundo lugar, mudamos a vírgula para esquerda, removendo zeros, ao tempo que, para compensar, avançamos com o exponente da base 10 para direita:</p>

<p>

$9 500 000 000 000 000 000. \cdot 10^0 \cdot km = 9 5. \cdot 10^{17} \cdot km = 9 5 \cdot 10^{17} \cdot km$

</p>

<p>E desta maneira, teremos transformado um número muito grande em notação científica.</p>

<p>Se, ao contrário, o número for muito pequeno procedemos de maneira análoga, mas sempre equilibrando a transformação: vírgula para direita e exponente da base 10 para esquerda. Usemos o número do raio atômico supracitado para conferir:</p>

<p>

$0.000 000 000 012 \cdot m =  0.000 000 000 012 \cdot 1 \cdot m = 0.000 000 000 012 \cdot 10^0 \cdot m = 12 \cdot 10^{-12} \cdot m$

</p>

</div>

## 6. Álgebra de Notação Científica
Ao longo desta disciplina de Física Geral vamos nos deparar com equações envolvendo grandezas físicas. Para resolver essas equações devemos substituir os valores de tais grandezas físicas, tipicamente números em formato de notação científica, e operar esses números. Podemos dividir essas operações em basicamente três: (1) a multiplicação e divisão, (2) a exponenciação e (3) a soma e subtração.

### 6.1 Mutiplicação e divisão de números em notação científica
Entendamos como multiplicar e dividir números em notação científica através de exemplos.

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>Calcule a área $A$ de uma fazenda a partir dos lados, de valores 2 10³ m e 3 10² m respectivamente.</b>

<p>

$A = comprimento \cdot comprimento = (2 \cdot 10^3 \cdot m) \cdot (3 \cdot 10^2 \cdot m) = 2 \cdot 10^3 \cdot m \cdot 3 \cdot 10^2 \cdot m = 2 \cdot 3 \cdot 10^3 \cdot 10^2 \cdot m \cdot m = 6 \cdot 10^5 \cdot m^2$

</p>

</div>

Observe que, na operação do exemplo anterior, houve três detalhes importantes. Primeiro, é usar sempre os parênteses na hora de insertar um valor numa equação. Segundo, é usar a propriedade comutativa, pois a ordem dos fatores (fatores são números, ou unidades, que multiplicam ou dividem) não altera o produto. Isso é fundamental para juntar entes similares para depois operar entre eles. Nesse caso, foi juntado números com números, bases com bases e unidades com unidades, mantendo assim a estrutura de um número em notação científica. E terceiro, usar a propriedade matemática em que os exponentes somam quando a base é a mesma.

Procedamos agora com a divisão.

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>Uma única força $F$ de 2 10³ N é aplicada sobre um corpo de massa $m$ igual a 3 10² kg, qual seria o valor da aceleração $a$ que o corpo experimenta?</b>

<p>

$a = \frac{F}{m} = \frac{(2 \cdot 10^3 \cdot N)}{(3 \cdot 10^2 \cdot kg)} = \frac{2 \cdot 10^3 \cdot kg \cdot \frac{m}{s²}}{3 \cdot 10^2 \cdot kg} = \frac{2}{3} \cdot \frac{10^3}{10^2} \cdot \frac{m}{s^2} = \frac{2}{3} \cdot 10^3 \cdot 10^{-2} \cdot \frac{m}{s^2}$

</p>

</div>

Neste caso, o detalhe principal foi que uma operação de divisão, no que respeita às bases de 10, se converteu numa operação de multiplicação, na qual os exponentes de bases iguais somam. Desta maneira, percebemos que dividir equivale a multiplicar pelo inverso do número que está no denominador. E o inverso de um número, ou de uma unidade, pois se comporta igual, construí-se mudando o sinal do seu exponente. De maneira geral, seja $x$ um número qualquer, o inverso de $x$ será:

$$\frac{1}{x} = x^{-1}$$

Insistimos em que esta mesma operação de tomar o inverso, feito com números, pode ser também feita para as unidades. Vejamos com outro exemplo.

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>Calcule a frequência ($f$) a partir do comprimento de onda ($\lambda$) de valor $2 \cdot 10^{-2}m$.</b>

<p>

$f = \frac{1}{\lambda} = \frac{1}{2 \cdot 10^{-2}m} = \frac{1}{2}\frac{1}{10^{-2}}\frac{1}{m} = 0,5 \cdot 10^2 \cdot m^{-1} = 5 \cdot 10^1 \cdot m^{-1} = 50 Hz$

</p>

<p>

Hertz, abreviado como $Hz$, é o nome que se da às unidades $m^{-1}$, as quais são unidades de frequência.

</p>

</div>

### 6.2 Exponenciação de números em notação científica
Entendamos também a operação de exponenciação através de um exemplo típico de Física.

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>Calcule a energia cinética ($K$) de um objeto, a qual depende da massa ($m$) da partícula e de sua velocidade ($v$) ao quadrado. Suponha para isso que $m$ vale $2 \cdot 10^3 kg$ e que $v$ vale $3 \cdot 10^2 m/s^2$.</b>

<p>

$K = \frac{1}{2} m v^2 = \frac{1}{2} (2 \cdot 10^3 \cdot kg) \cdot (3 \cdot 10^2 \frac{m}{s})^2 = \frac{1}{2} (2 \cdot 10^3 \cdot kg) \cdot (3^{1 \cdot 2} \cdot 10^{2 \cdot 2} \frac{m^{1 \cdot 2}}{s^{1 \cdot 2}}) = \frac{1}{2} (2 \cdot 10^3 \cdot kg) \cdot (3^2 \cdot 10^4 \frac{m^2}{s^2}) = \frac{1}{2} \cdot 2 \cdot 3^{2} \cdot 10^3 \cdot 10^4 \frac{kg \cdot m^{2}}{s^{2}} = 9 \cdot 10^7 \frac{kg \cdot m^{2}}{s^{2}} = 9 \cdot 10^7 J$

</p>

<p>Os Joules (J) é a unidade associada à grandeza física energia.</p>

</div>

Observe-se o detalhe principal na operação de exponenciação: ao elevar um número em notação científica a qualquer exponente, o exponente de todos os fatores multiplica pelo exponente externo.

### 6.3 Soma e subtração de números em notação científica
A última operação entre números em notação científica é a soma/subtração. Já que um número em notação científica está composto por, no máximo, 3 fatores (o número mesmo, a base 10 e a unidade) a maneira de proceder numa soma/subtração consiste em procurar ou, se for possível criar, fatores em comum aos dois somandos e deixar eles em evidência. Vejamos com um exemplo. 

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>A equação fundamental da hidrostática relaciona a pressão ($p$) a uma profundidade específica, com o valor dessa profundidade ($h$), com a densidade do fluido ($\rho$), com a aceleração da gravidade ($g \approx 10 m/s^2$) e com a chamada pressão de referência ($p_0$). Supondo os seguintes valores: $h = 1 m$ (de profundidade), $\rho = 10^3 kg/m^3$ (a densidade da água), e $p_0 = 10^5 Pa$ (a pressão atmosférica no nível do mar), qual seria o valor da pressão?</b>

<p>

$p = p_0 + \rho \cdot g \cdot h = 10^5 \cdot Pa + 10^3 \frac{kg}{m^3} \cdot 10 \frac{m}{s^2} \cdot 1 \cdot m = 10^5 \cdot Pa + 10^4 \frac{kg}{m s^2} = 10^5 \cdot Pa + 10^4 \cdot Pa = (10^5 + 10^4) Pa = (10 \cdot 10^4 + 10^4) Pa = (10 + 1) 10^4 \cdot Pa = 11 \cdot 10^4 \cdot Pa$

</p>

</div>

Observe-se, portanto, como foi explicado anteriormente que, na hora de somar devemos ir procurando, ou criando, fatores em comum e deixando-os em evidência.

Finalmente, para efetuar uma subtração de números em notação científica devemos lembrar do elemento oposto da soma. Seja $y$ qualquer número, ou unidade, seu oposto será $-y$. Dito isso, devemos entender que a subtração não é mais que somar um número com seu oposto:

$$x - y = x + (-y)$$

e, portanto, a operação de subtração entre dois números é análoga à soma. E sendo assim, todas as contas explicadas anteriormente seriam as mesmas com a única mudança de substituir um símbolo de somar pelo de subtrair e ter que efetuar dita subtração final. 

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>A energia total de uma partícula é a soma da energia cinética com a energia potencial. Suponhamos que a energia cinética ($K$) tenha um valor de $2 \cdot 10^3 J$ enquanto que a energia potencial ($U$) tenha um valor de $-3 \cdot 10^2J$. Qual será a energia total ($E$):</b>

<p>

$E = K + U = (2 \cdot 10^3 \cdot J) + (-3 \cdot 10^2 \cdot J) = (2 \cdot 10^3 -3 \cdot 10^2)J = (2 \cdot 10 \cdot 10^2 -3 \cdot 10^2)J = (2 \cdot 10 -3)10^2 \cdot J = 17 \cdot 10^2 \cdot J$

</p>

</div>

## 7. Prefixos
A grande vantagem do Sistema Internacional de Unidades, e provavelmente o que explica sua boa acolhida mundial, é o fato de usar um conjunto de unidades relacionadas entre elas através de fatores de 10, o qual, por sua vez, simplifica a transformação de unidades através de fatores de conversão. Algumas destas potências de 10 são escolhidas para, a partir delas, derivar unidades maiores ou menores. ## 8. Transformação de unidades
Dado que, como vimos anteriormente, dispomos de vários sistemas de unidades e, por sua vez, dentro de cada sistema, encontramos uma série de prefixos, ou seja, subunidades de cada unidade, é imprescindível saber como **transformar unidades** de um número. Aqui é onde entram os chamados **fatores de conversão**, que será a técnica que usaremos para transformar unidades, a qual está baseada no uso da notação científica e, especificamente, nas bases de 10.

A técnica dos fatores de conversão consiste em procurar fatores que relacionem unidades diferentes, as quais queremos transformar. Por exemplo, se quisermos efetuar uma transformar entre unidades de comprimento, por exemplo, entre $km$ e $m$, precisamos de uma relação simples entre essas unidades. A relação mais simples e conhecida é:

$$1 \cdot km = 10^3 \cdot m$$

Agora, tal equivalência entre as unidades deve ser convertida em um fator de conversão, o qual é o que usaremos para efetivar a transformação. Já que o número 1 é o neutro da multiplicação, podemos então proceder assim a partir da igualdade anterior:

$$1 \cdot 1 \cdot km = 10^3 \cdot m$$

e isolando o número 1 que foi colocado à esquerda,

$$1 = \frac{1 \cdot km}{10^3 \cdot m}$$

A parte que restou à direita do igual se chama **fator de conversão direto** e, como vemos, ele vale a unidade. Isso será muito importante porque quer dizer que, como o fator de conversão vale a 1, podemos insertar ele nas contas em qualquer momento e transformar.

Da mesma maneira em que de uma relação entre unidades foi extraído um fator de conversão direto, pode-se também extrair o chamado **fator de conversão inverso**. Para isso, desta vez, insertamos o número 1 à direita do igual:

$$1 \cdot km = 10^3 \cdot m \cdot 1$$

e isolamos ele,

$$1 = \frac{10^3 \cdot m}{1 \cdot km}$$

Finalmente, agora sim podemos nos focar mesmo na transformação. Caso quisermos transformar 1 km para m devemos usar o fator de conversão inverso, pois só este é quem faz cortar as unidades de km:

$ 1 \cdot km = 1 \cdot km \cdot 1 = 1 \cdot km (\frac{10^3 \cdot m}{1 \cdot km}) = 1 \cdot km \frac{10^3 \cdot m}{1 \cdot km} = \frac{10^3 \cdot m \cdot km}{km} = 10^3 \cdot m$

Observe como o fator de conversão foi insertado nas contas sem alterar as mesmas já que ele vale a unidade. Observe também que em caso de efetuar a transformação inversa, ou seja, transformar 1 m para km, devemos usar o fator de conversão inverso, pois ele que faz apagar as unidades de m:

$ 1 \cdot m = 1 \cdot m \cdot 1 = 1 \cdot m (\frac{1 \cdot km}{10^3 \cdot m}) = 1 \cdot m \frac{1 \cdot km}{10^3 \cdot m} = \frac{km \cdot m}{10^3 \cdot m} = 10^{-3} \cdot km$


Como vemos, a técnica dos fatores de conversão converte unidades de maneira limpa e elegante, pois não precisamos sair das contas, ao contrário da técnica da regra de três. Entretanto, até agora, todo o anterior poderia ter sido feito usando esta segunda técnica, deixando em dúvida a utilidade da primeira. Na verdade, o verdadeiro potencial da técnica dos fatores de conversão ainda não foi explicado, pois é aí onde dita técnica se destaca, superando claramente à regra de três. Vejamos alguns exemplos desse potencial.
