## 1. Introdução
No capítulo 4, que tratava sobre Cinemática, iniciamos o estudo do movimento. Naquele momento explicamos que o ramo da Física que estuda o movimento é a Mecânica e ele se compõe da Cinemática e da **Dinâmica**. Neste capítulo, portanto, estudamos a parte que falta para entender completamente o fenômeno do movimento.

Novamente, igual que fizemos ao introduzir a Cinemática, devemos simplificar conceitos para facilitar a abordagem. A Dinâmica que estudaremos será a chamada Dinâmica Translacional Clássica da Partícula, na qual também devemos tratar aos corpos como se fossem pontos para assim evitar a rotação e, como consequência, apenas abordaremos a translação desses corpos pontuais.

Dinâmica provém da palavra grega *dinamo*, a qual tem a ver com explosão. Efetivamente, a Dinâmica estuda como o movimento 'explode', no sentido em como ele é gerado. Essa questão é fundamental, mas até agora não tínhamos resposta. Isso foi devido a que a Cinemática não responde quem gerou o movimento nem o por quê. A tarefa neste capítulo será responder a essas questões que faltam para descrever completamente o movimento.

Como agora sabemos, o movimento se divide em dois: o começo (Dinâmica) e o desenvolvimento (Cinemática), áreas que efetivamente se estudam de maneira independente. Entretanto, nessa explosão do movimento, alguma informação deve ser extraída para que, a partir, o movimento se desenvolva ao longo do tempo. Essa informação está embutida na aceleração. A aceleração é, de fato, o nexo de união entre ambas as áreas, a Cinemática e a Dinâmica.

As grandezas físicas que estudávamos em Cinemática eram, principalmente, o tempo, a posição, a velocidade e a aceleração. Agora, em Dinâmica, além da aceleração, emergem duas novas grandezas físicas: a **massa** e a **força**, a partir de uma expressão fundamental conhecida como **Segunda Lei de Newton**.

Em Cinemática definíamos duas grandezas físicas principais: o tempo e a posição, e o resto eram derivadas dessas a partir de operações matemáticas simples como subtrações e divisões, ou a partir de operações matemáticas mais complexas como a derivada ou a integral. Todo o arcabouço da Cinemática foi construída dessa maneira. Agora, faremos algo parecido, só que, ao invés de partir de definições de grandezas físicas, partiremos da Segunda Lei de Newton. E, a partir da mesma, usando ela como base, será construída toda a Dinâmica através de, agora sim, ferramentas lógicas.

Curiosamente, a Segunda Lei de Newton não é uma expressão racional, no sentido em que não pode ser deduzida a partir da lógica. A Segunda Lei de Newton é uma expressão empirista, pois ela é deduzida a partir da experiência. Isso é devido a que, como veremos, o conceito de força que usaremos surge a partir de uma média das interações dadas por um número enorme de partículas (átomos ou moléculas) das quais os corpos macroscópicos envolvidos se compõem. E, quando isso acontece, no momento em que ao invés de tratar com partículas puntuais sem nenhuma natureza interna (definição de ponto matemático tal e como foi usado em Cinemática), tratamos de partículas reais (átomos ou moléculas) a pureza e rigor matemático, que a priori sempre existe, se dilui no caos. É partir daí, onde a lógica matemática não é tão útil quanto o são nossos sentidos.

A Segunda Lei de Newton é a equação mais importante de Dinâmica. Entretanto, ela não a única. De fato, os pilares de Dinâmica estão baseados em três leis: a Primeira, a Segunda e a Terceira Lei de Newton. Cada Lei trata de um assunto diferente e independente, o qual não é consequência da Lei anterior. Então, é só por motivos históricos, que as tais leis foram enumeradas dessa maneira. Entretanto, apenas para facilitar a abordagem da Dinâmica, trataremos primeiro a Segunda Lei e posteriormente as outras duas.

## 2. Conceito de força
Como temos visto anteriormente a força é a origem do movimento e, sendo assim, o conceito de força, junto com a natureza da matéria, sempre tem intrigado aos filósofos ao longo da história. Curiosamente, o conceito de força está estreitamente ligado à matéria, ou mais profundamente, às partículas das quais se compõe a matéria. Isto é assim porque cada partícula tem uma natureza específica e é essa sua natureza que define como será a força que senta com outras partículas similares. Na verdade, quando nos adentramos profundamente na estrutura da matéria o próprio conceito de partícula se dilui, pois a mesma não é tratada realmente como uma partícula, mas como um campo, os quais interagem com outros campos que representam outras partículas. E o conceito de força, realmente, surge da interação dos campos que as partículas formam. Essa visão da matéria está explicada pela Teoria Quântica de Campos, a qual fica fora do escopo deste livro. Entretanto, é importante entender o básico: a força surge da interação entre partículas da mesma natureza, sendo algo inerente a elas.

O desenvolvimento atual da Física determina que existem 4 forças fundamentais, chamadas **Forças Fundamentais da Natureza**, a saber:
1. Força eletromagnética, responsável dos fenômenos elétricos e magnéticos.
2. Força nuclear forte, responsável de que os núcleos dos átomos e moléculas não se separem.
3. Força nuclear fraca, responsável da desintegração de certas partículas.
4. Força gravitacional, responsável pela atração das massas.

Das quatro forças fundamentais, as três primeiras emergem da interação entre as chamadas **partículas elementais** da Natureza, ou seja, as menores partículas que existem, pois, supostamente, se consideram indivisíveis. Já a quarta força, a força gravitacional possui uma Natureza um tanto diferente. É verdade que existe uma importante teoria física chamada Teoria da Gravitação Quântica que pretende visualizar a força gravitacional também como uma interação entre partículas elementais. Essas teorias são importantes porque permitiriam entender a Física da maneira unificada, através de uma Teoria do Tudo. Entretanto, para que uma teoria da quantização do campo gravitacional tenha evidência empírica, deveria ser detectada a partícula responsável de transmitir a força gravitacional, o chamado gravitão, e ainda não se tem evidências de que ele realmente exista.

Todas as forças são importantes porque a partir delas podemos explicar os fenômenos da Natureza. Entretanto, na Física do dia a dia, da qual trata este livro, apenas sentiremos as consequências da força gravitacional, da qual emergem as chamadas **forças a distância** e da força elétrica, da qual emergem as chamadas **forças de contato**.

### 2.1 Forças a distância
A força gravitacional surge da chamada **Lei de Gravitação Universal**, a qual estabelece que o módulo da força $F^{grav}$ com a qual uma massa $m_1$ atrai a uma massa $m_2$ vem dada por:

$$\tag{*} F^{grav} = G \frac{m_1 m_2}{d^2}$$

, onde $d$ é a distância que separa as massas e $G$ é chamada constante de gravitação universal de valor $6.67\cdot 10^{-11} \frac{m^3}{kg\cdot s^2}$.

A Lei de Gravitação Universal indica que todas as massas se atraem, mas para que essa atração seja considerável as massas em jogo devem ter valores grandes, já que a constante de gravitação universal é muita pequena. Efetivamente, todos os objetos a nosso redor se exercem forças de gravidade entre eles, mas essa força é geralmente desprezível. Entretanto, quando a massa de um corpo é relativamente grande comparada com o valor da constante, como pode ser a massa do planeta Terra, e a distância que separa os corpos é pequena, como acontece com os objetos ao nosso redor, situados a centímetros ou metros da superfície terrestre, a força de gravidade se manifesta, sendo esta a chamada **força peso**. De fato, se na expressão anterior tomamos $m_1$ como a massa da Terra ($M$), $m_2$ como a massa de um objeto qualquer ($m$) e a distância entre os centros dos corpos como o raio da Terra ($R$):

$$\tag{*} F^{grav} = G \frac{m_1 m_2}{d^2} = G \frac{M m}{R^2} = m\frac{G M }{R^2} = mg$$

, onde $g$ é a chamada **aceleração da gravidade**, cujo valor, para um corpo nas proximidades da superfície terrestre, é aproximadamente:

$$ g \approx 10 \frac{m}{s^2}$$

Desta maneira, como vemos, a força peso ($P$) é uma força fundamental da natureza, a força de gravidade, que surge como consequência da Lei de Gravitação Universal.

$$\tag{*} P = mg$$

ou, de maneira geral,

$$\tag{*} \vec{P} = m\vec{g}$$

já que a aceleração da gravidade realmente tem um sentido e direção (todos os corpos caem para o centro da Terra) e a força peso, portanto, também.

### 2.2 Forças de contato
A força elétrica surge da chamada **Lei de Coulomb**, a qual estabelece que o módulo da força elétrica $F^{ele}$ com a qual uma carga $q_1$ atrai ou repele uma carga $q_2$, medidas nas unidades de **Coulombs** ($C$), vem dada por:

$$\tag{*} F^{ele} = k \frac{q_1 q_2}{d^2}$$

, onde $d$ é distância que separa as cargas e $k$ é uma constante de proporcionalidade de valor aproximado $9\cdot 10^9 N\cdot m^2/C^2$.

Como vemos, a expressão da força elétrica tem o mesmo formato que força gravitacional. Entretanto, há algumas diferenças importantes. A primeira é que a força gravitacional é sempre atrativa porque não fazem sentido massas negativas. Já a força elétrica pode ser atrativa ou repulsiva, e isso vai depender da carga dos objetos envolvidos. Particularmente, um objeto pode estar carregado positivamente quando os átomos ou moléculas das quais se compõe estiverem com falta de elétrons ou pode estar carregado negativamente quando os átomos ou moléculas das quais se compõe estiverem com excesso de elétrons. A segunda diferença está no valor das constantes. Enquanto a constante de gravitação $G$ é muito pequena, a constante $k$ é enorme. Isso implica que a força elétrica é muito maior que a força gravitacional. Esse efeito está na essência de fenômenos tão simples como o fato de que possamos segurar qualquer objeto e ainda transportá-lo para outro lugar.

Da mesma maneira que a força gravitacional, a força elétrica é também uma força exercida a distância. Entretanto, a grande maioria dos objetos ao nosso redor estão em estado eletricamente neutro e, sendo assim, a força elétrica não se manifesta. A força elétrica emerge, porém, quando a distância entre os objetos é tão pequena que quem entra em jogo não é mais o objeto em si, se não as partículas elementares das quais se compõe. De fato, quando um objeto sólido é aproximado a outro observamos que, uma vez que entram **em contato**, não podemos mais aproximar os corpos. A causa disso, no fundo, é que os elétrons dos átomos/moléculas dos objetos envolvidos exercem entre eles forças elétricas repulsivas. A soma de todas essas forças repulsivas exercidas por uma quantidade enorme de elétrons cria uma força resultante que é quem evita que os corpos se "fusionem", entrando um dentro do outro.

É dessa maneira como emerge a chamada **força normal**, a qual é uma força que tipicamente exerce uma superfície contra um corpo que é empurrado por outra força contra a mesma. A força normal emerge como resultante da repulsão de um grande número de elétrons, que evitam que um objeto entre dentro da superfície de outro corpo. A força normal tem esse nome porque ela é sempre perpendicular à superfície e para fora dela. 

É importante insistir em que a força de normal só emerge quando, macroscopicamente, há contato entre as superfícies dos corpos envolvidos e, por isso, se diz que se trata de uma força de contato. Entretanto, como estamos explicando, a um nível microscópico esse contato realmente não existe, pois a força de repulsão tende a infinito quando a distância entre as partículas envolvidas, neste caso os elétrons dos objetos "em contato", tende a zero.

### 2.3 Forças de coesão
A Natureza elétrica dos objetos, e particularmente a repulsão eletrônica, nos ajuda a entender qualquer força que consista em um **empurrão** através de um contato macroscópico. Mas que acontece quando exercemos um **puxão**, como quando puxamos manualmente de algum objeto ou quando exercemos tal puxão por meio de alguma corda através de uma **força de tensão**? 

A resposta está novamente na estrutura da matéria. De fato, os átomos/moléculas dos objetos sólidos se estruturam através de ligações químicas, as quais não são facies de quebrar. Essas ligações são precisamente as que fornecem tal solidez à matéria e a que explica o fato que possamos puxar de objeto sem que, até certo limite, ele quebre.

## 3. Segunda Lei de Newton
A Segunda Lei de Newton vem dada pela seguinte expressão:

$$\boxed{\tag{*}\Sigma \vec{F} = m \cdot \vec{a}}$$

onde $\Sigma \vec{F}$ é a resultante, ou soma, das forças que atuam num corpo de massa $m$, as quais lhe produzem uma aceleração $\vec{a}$. O símbolo $\sigma$ é habitualmente usado em Matemática para denotar uma soma. Neste caso $\Sigma \vec{F}$ consistiria em somar vetorialmente todas as forças que atuam no corpo, alías, $\vec{F_1}+\vec{F_2}+\vec{F_3}+...$

Como sabemos do primeiro capítulo as unidades da massa no S.I. são os $kg$ e, como sabemos da capítulo 4, as unidades da aceleração no S.I. são $m/s^2$. Portanto, da própria Segunda Lei as unidades da força vem do produto das unidades da massa com as unidades da aceleração, ou seja, $kg\cdot \frac{m}{s^2}$, as quais se denominam Newtons, abreviado pela letra $N$ maiuscula.

A Segunda Lei é uma equação causal porque estabelece uma relação entre uma causa: a soma das forças que atuam no corpo, e uma consequência: a aceleração que o corpo experimenta. Ela também é uma equação vetorial e, portanto, não é só uma equação de números (módulos), mas também de direções e sentidos. De fato, se a massa tivesse valor unidade ($1kg$), a segunda Lei de Newton simplificaria a:

$$\Sigma \vec{F} = a$$

e, nesse caso, a aceleração que o corpo sofreria seria igual à força resultante em módulo, direção e sentido. É muito lógico que isso aconteça, pois por experiência sabemos que o corpo se move (acelera) para o lugar para onde ele é empurrado.

Isso também acontece no caso geral onde o valor da massa é diferente de um, pois como vimos no capítulo sobre vetores, multiplicar um escalar por um vetor, neste caso, só resulta numa mudança do módulo, já que no existem massas negativas. Sendo assim, multiplicar a massa pelo vetor $\vec{a}$ só resulta em esticar o vetor aceleração (quando $m>1$) ou em reduzi-lo (quando $m<1$).

Como também sabemos, todo vetor pode ser colocado em função das suas componentes. Ou seja, no caso de um arranjo em 2D, a Segunda Lei também poderia ser escrita assim:

$$\tag{*}\Sigma \vec{F_x} + \Sigma \vec{F_y} = m \cdot (\vec{a_x} + \vec{a_y})$$

ou, em consequência:

$$\tag{*}\Sigma F_x\hat{i} + \Sigma F_y\hat{j} = m \cdot (a_x\hat{i} + a_y\hat{j})$$

E, finalmente, podemos dividir essa equação em outras novas, as quais representariam sua decomposição vetorial. Para isso, multiplicamos essa equação pelo vetor unitário $\vec{i}$ e, por outro lado, pelo vetor unitário $\hat{j}$. Um pouco de álgebra vetorial usando o produto escalar de vetores unitários deveria nos levar a que:

$$\boxed{\tag{*}\Sigma F_x = m \cdot a_x}$$

$$\boxed{\tag{*}\Sigma F_y = m \cdot a_y}$$

Estas equações são a Segunda Lei de Newton, porém em formato escalar e elas são importantes porque representam uma alternativa a usar Segunda Lei de Newton no seu formato vetorial. É interessante observar que sempre podemos decompor uma equação vetorial em várias equações escalares, tantas quanto a dimensão do espaço problema. Trabalhar com equações escalares é bastante mais fácil que trabalhar com equações vetoriais e, por isso, na hora de resolver problemas essa será a abordagem preferencial. Entretanto, nem tudo são vantagens, pois passar de usar equações vetoriais a escalares desemboca no *trade-off* inevitável de ter que usar várias equações. De fato, essa transformação sempre desemboca num sistema de equações, o qual pode também ser representado matricialmente:

$$
\begin{pmatrix}
   \Sigma F_x \\ \Sigma F_y
\end{pmatrix}=
\begin{pmatrix} 
   m & 0 \\
   0 & m
\end{pmatrix}
\begin{pmatrix} 
   a_x \\ a_y 
\end{pmatrix}
$$

É claro que ao passar do formato vetorial ao formato escalar chegamos a um sistema de equações que são **linearmente independentes**, pois todas as equações escalares nos aportam informação relevante e complementar. É importante entender que, como consequência de independência linear das equações, o tratamento do que acontece ao longo de cada eixo de coordenadas é independente. Portanto, vamos a poder estudar as leis de Newton de maneira independente para cada eixo.

Muitos problemas envolvendo forças se resolvem usando o mesmo roteiro, o qual pode se tornar de grande ajuda, sobre tudo quando não temos nem ideia de como proceder. Descrevamos a seguir os passos porque usaremos o mesmo com frequência.
1. **Esboçar um desenho do problema.** Um desenho ajuda a visualizar o que está acontecendo e nos forçar a aplicar a lógica para desenvolver os passos seguintes
2. **Levantar um diagrama de corpo livre.** Um diagrama de corpo livre é um desenho simplificado no qual os corpos envolvidos são isolados do seu entorno. Sobre esses corpos, tantos quanto tenhamos no problema, é onde se aplicam as forças, as quais também devemos colocar.
3. **Escolher um referencial.** Em algum momento do problema é necessário escolher o sinal dos vetores e para isso não há outra opção que usar um referencial. Existem infinitos referenciais, mas devemos escolher aquele que facilite a resolução do problema.
4. **Descompor as forças.** Nos problemas que envolvem vetores é muito conveniente trabalhar com as componentes deles, pois geralmente é mais fácil trabalhar com escalares que com números. Cada forçar que precisar ser descomposta gerará 2 equações, uma envolvendo o seno e outra o cosseno.
5. **Aplicar as Leis de Newton.** Finalmente aplicamos a Segunda Lei de Newton a cada corpo. Caso houver mais de um corpo envolvido deveremos também usar a Terceira Lei.

## 4. Conceito de massa
Podemos usar a Segunda Lei de Newton para entender o que realmente significa a massa. Suponhamos que sobre um corpo de massa $m$ qualquer atua apenas uma força, a qual chamaremos de $F$. Nesse caso a Segunda Lei de Newton simplificaria a:

$$\vec{F} = m \cdot \vec{a}$$

Mas já que apenas existe uma força, e já que, como sabemos, o vetor aceleração tem que ir no mesmo sentido e direção do vetor $\vec{F}$, sempre podemos escolher o referencial com o semi-eixo positivo de X orientando no sentido e direção dessa força $F$. E sendo assim, os vetores não são mais necessários, com o qual teríamos que:

$$F = m \cdot a$$

E, finalmente, para nosso propósito de entender o significado da massa, isolemos a aceleração:

$$a = \frac{F}{m}$$

Analisemos agora esta equação do ponto de vista funcional. De esse ponto de vista teríamos uma função que depende de duas variáveis, ou seja, $a=f(F,m)$. Já que, no capítulo de funções apenas estudamos funções de uma variável, estudemos que acontece se apenas uma dessas grandezas físicas é variável, deixando para isso a outra como constante. Comecemos primeiro escolhendo um experimento no qual a massa de um corpo é a mesma, e sobre esse corpo atuam forças de diferente magnitude com o intuito de estudar a aceleração que o corpo experimenta. Já que, dessa maneira, a equação que fica seria uma equação linear,

$$a = \frac{F}{m} = \frac{1}{m}F$$

e seu gráfico seria uma linha reta que parte da origem e com inclinação $1/m$.

Observe-se como, supondo uma massa $m$ constante, entre $a$ e $F$ existe o que é chamada uma **relação diretamente proporcional**, já que, dada uma força $F$ qualquer, se esta aumenta, então o valor da aceleração aumenta na mesma proporção. Por exemplo, se na equação anterior insertamos uma nova $F'$ dobre da anterior, o resultado é uma nova aceleração que também é o dobro da anterior.

$$a' = \frac{1}{m}F' = \frac{1}{m}(2F) = 2\frac{1}{m}F = 2 a$$

Desta relação podemos finalmente entender o significado da massa. Desse ponto de vista a massa é algo que dificulta o movimento, pois quanto maior for, mais difícil será para um corpo começar a se mover (acelerar). Então, a massa é a resistência que um corpo oferece a ser movimentado.

Isso é algo lógico para nós e estamos completamente habituados a isso. Por exemplo, não duvidamos nem um pouco na hora de querer mudar a posição de qualquer objeto pequeno como um celular, mas rapidamente descartamos que podamos pegar um objeto maior, como um carro ou um caminhão, e poder mudar sua posição aplicando a mesma força.

## 5. Primeira Lei de Newton
A segunda Lei de Newton confirma que um corpo acelera se existe uma força resultante. Entretanto, que acontece se essa resultante é zero? Nesse caso a Segunda Lei ficaria assim:

$$0 = m \cdot \vec{a}$$

e, já que a massa não pode ser zero (se for assim não existiria o corpo problema), o vetor aceleração seria zero. E, como sabemos do capítulo de Cinemática, quando a aceleração é zero se trataria de um Movimento Retilíneo e Uniforme. Isso é precisamente o enunciado da Primeira Lei de Newton: "Se sobre um corpo não atuam forças, ou atuam mas elas se cancelam, o corpo seguirá um Movimento Retilíneo e Uniforme." Escrito analaticamente:

$$\boxed{\tag{*}\Sigma \vec{F} = 0 \Longleftrightarrow \vec{a}=0}$$

Da maneira em que foi introduzida, parece que a Primeira Lei é consequência da Segunda. Realmente, a Primeira Lei é um caso particular da Segunda. Entretanto, ela é convenientemente separada em outra lei independente porque a partir da mesma pode se demostrar que o conceito de **referencial inercial**. Em capítulos anteriores aprendemos que o referencial é necessário e também que ele é uma escolha. Porém, naquele momento apenas foram usados referenciais fixos. Uma pergunta que teria cabido naquele momento era se poderíamos escolher referenciais em movimento com uma determinada velocidade ou ainda como uma determinada aceleração. Isso é o que precisamente determina o conceito de referencial inercial. As Leis de Newton seriam diferentes se usássemos referenciais acelerados. Por exemplo, teríamos que usar forças que realmente não existem como são as chamadas forças centrífugas. Entretanto, como consequência da Primeira Lei, nos referenciais inerciais, garante-se que as Leis de Newton são as mesmas em todos eles. Os referenciais inerciais são referencias não acelerados (velocidade constante) e serão eles, particularmente os de velocidade constante igual a zero os que usaremos neste livro. 

A Primeira Lei é particularmente importante no ramo da Engenheira Civil. De fato, na hora de construir uma casa é fundamental, claro, que ela não acelere translacionalmente nem rotacionalmente. E para isso, na verdade para a primeira condição (pois as rotações estão fora do nosso escopo), as forças que atuam sobre as paredes de uma casa devem ser zero. É por isso que a Primeira Lei é conhecida como **condição de equilíbrio**.

$$\boxed{\tag{*}\Sigma F_x = 0}$$

$$\boxed{\tag{*}\Sigma F_y = 0}$$

Considerar a condição de equilíbrio é fundamental em áreas como Engenheira Civil. De fato, para construir os pilares de uma casa ou de um prédio, e ter certeza que nada vai desabar, duas condições devem ser satisfeitas:
- As somas das forças devem ser zero ($\Sigma \vec{F} = 0$), para que os pilares não acelerem traslacionalmente.
- As somas dos torques devem ser zero ($\Sigma \vec{\tau} = 0$), para que os pilares não acelerem rotacionalmente.

O símbolo $\vec{\tau}$ denota um torque, que tem a ver com quão uma força, em função de onde ela se aplica, é capaz de fazer girar um objeto extenso. Mas, como já avisamos, estudar a rotação está fora do escopo deste livro.

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>Um engenheiro deseja colocar um semáforo de $10 kg$ a uma certa altura no meio da pista. O único pilar de apoio é uma parede no lateral da estrada. Como faria isso e qual deveria ser o módulo das forças que teria que aplicar no semáforo para satisfazer a condição de equilibrio?</b>

Se nenhuma força extra for aplicada o semáforo estaria em queda livre devido à força peso. Podemos então usar um arame do ponto A ao ponto B para segurar o semáforo. Esse arame exerceria uma força de tensão ($\vec{T}$), a qual formaria um ângulo $\theta$ com a horizontal. Entretanto, apenas com isso o semáforo ligado ao arame giraria em direção à parede, pois para lá iria a força resultante do peso ($\vec{P}$) e a tensão do arame. Para evitar isso é conveniente acrescentar uma haste do ponto B até o ponto C que exerça uma força $\vec{F}$ que cancele a soma das forças anteriores e o sistema com um todo esteja equilibrado.

Dado esse sistema, que potencialmente poderia equilibrar o semáforo, procuremos agora o valor das forças $\vec{T}$ e $\vec{F}$ para que realmente haja condição de equilíbrio. Para isso, escolhemos um referencial, descompomos as forças que estiverem fora dos eixos e aplicamos a Primeira Lei de Newton.

No nosso caso, a escolha feita do referencial indica que apenas a força $\vec{T}$ precisa ser descomposta. Aplicando as definições de seno e cosseno concluímos que:

$$T_x = T \cos\theta$$
$$T_y = T \sin\theta$$

E, finalmente, aplicando a Primeira Lei de Newton no eixo X e depois no eixo Y, e levando em consideração que o peso é o produto da massa pela aceleração da gravidade, teríamos que:

$$T_x - F = 0$$
$$T_y - mg = 0$$

Observe-se que estas duas equações junto com as outras duas anteriores formaria um sistema de 4 equações linearmente independentes, onde só pode haver 4 incógnitas, que seriam $T$, $F$, $T_x$ e $T_y$. Já que as duas últimas não nos interessam, podemos transformar esse sistema de 4 equações e 4 incógnitas em outro de 2 equações com 2 incógnitas:

$$T \cos\theta = F$$
$$T \sin\theta = mg$$

Finalmente, da segunda equação obtemos o valor do módulo da tensão:

$$T = \frac{mg}{\sin\theta }$$

e insertando esse valor na primeira equação obtenemos o valor do módulo da haste:
$$F = T \cos\theta = \frac{mg}{\sin\theta } \cos\theta = \frac{mg}{\tan\theta }$$

A primeira Lei de Newton também é conhecida como **Lei de inércia**. Inércia quer dizer que se um corpo não é incomodado, no sentido em que as forças que atuam sobre ele são zero ou se cancelam, então o corpo continua fazendo o que estava fazendo; o corpo segue sua inércia, ou seja, um Movimento Retilíneo e Uniforme. A lei de inércia, ou a Primeira Lei, é muito importante para explicar fenômenos do nosso dia a dia como porque nosso corpo vai para frente quando um ônibus freia ou por que tende a sair pela tangente quando o ônibus faz uma curva.

## 6. Terceira Lei de Newton

Os exemplos que estudamos até agora, nos quais aplicamos a Primeira e Segunda Lei de Newton, foram situações simples, nas quais o corpo de estudo era, digamos, isolado do resto do Universo. De fato, no exemplo típico, no qual alguém empurra um bloco, nosso foco era estudar apenas a aceleração que experimentava o bloco e não nos preocupávamos sobre de onde aparecia essa força e que poderia acontecer se, por exemplo, fosse uma pessoa quem empurrava esse bloco.

Isolar um corpo do resto do Universo é uma abordagem muitas vezes necessária para poder entender o problema. Entretanto, não é assim como funciona a Natureza. A Física é realmente bem mais complicada que os exemplos ideais que estudamos, pois na realidade tudo (*hole* en inglês) interage com tudo. A terceira Lei de Newton vem a tona precisamente para isso, para entendermos que não existem forças isoladas. 

<div style="padding: 15px; border: 1px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #8a6d3b;; background-color: #fcf8e3; border-color: #faebcc;">

<b>Estudemos o sistema Terra-Lua para entender a interação entre as partes do sistema e as acelerações que cada parte sofreria como consequência dessa interação.</b>

<p>

A Terra e a Lua formam um sistema, cujas partes permanecem ligadas devido à força de gravidade. A Terra e a Lua são realmente partes de um sistema maior, o sistema solar, cujas partes também interagem com o sistema Terra-Lua. Entretanto, por simplificação, suponhamos que o mesmo se encontra isolado do Universo. Ou seja, neste experimento mental analisamos o que acontece na Terra e na Lua como se eles fossem o único que existisse no Universo.

</p>
<p>

Nesse arranjo sabemos que, pela primeira Lei de Newton, se não atuassem forças sobre ela, a Lua seguiria um M.R.U. e escaparia pela tangente à sua trajetória circular ao redor da Terra. Sabemos que isso não acontece, e se não acontece é porque atua uma força, neste caso a força de gravidade, que é uma força de atração entre as massas dada pela Lei de Gravitação Universal. A Terra, portanto, exerce uma força de atração na Lua $\vec{F}_{TL}$. Da mesma maneira, pela Terceira Lei, a Lua exerce outra força de atração na Terra $\vec{F}_{LT}$.

</p>
<p>

Podemos esboçar um desenho de um instante qualquer da rotação da Lua ao redor da Terra e levantar um diagrama de corpo livre. Escolhendo um referencial que simplifique as contas, podemos aplicar a Segunda Lei ao longo do eixo X a cada um dos corpos envolvidos, a massa da Terra ($m_T$) e massa da Lua ($m_L$):

</p>

$$F_{LT} = m_T \cdot a_T$$

$$-F_{TL} = m_L \cdot a_L$$

<p>

, onde $a_T$ é a aceleração que sente a Terra e $a_L$ é a aceleração que sente a Lua. Pela Terceira Lei de Newton sabemos que $F_{LT} = F_{TL}$ e sendo assim

</p>

$$m_T \cdot a_T = - m_L \cdot a_L$$

<p>

E, finalmente, isolando o valor a aceleração que sente a Terra:

</p>

$$\frac{a_T}{a_L} = -\frac{m_L}{m_T}$$

<p>

Esta expressão contém muita informação a qual podemos analisar separadamente.
1. Tanto a Terra quanto a Lua estão aceleradas. Essas acelerações são **acelerações centrípetas** e elas surgem como consequência da força de gravidade, quem atua como força centrípeta. Tais acelerações são contrárias, o que provocaria que os astros se encontrassem em algum ponto intermédio da linha que os une. Entretanto, como sabemos isso não acontece já que a Lua se move com certa velocidade ao redor da Terra. Então, a Lua nunca vai cair para o centro da Terra. Ela de fato está num equilíbrio: por um lado a Lua quer sair pela tangente, devido à Primeira Lei de Newton, mas a aceleração centrípeta devido à força gravitacional que exerce a Terra evita que a Lua escape.
2. A massa da Terra é muito maior que a massa da Lua, resultando num quociente menor que 1 no lado direito da equação. Isso implica, no quociente do lado esquerdo, que a aceleração que sente a Lua deverá ser muito maior que a que sente a Terra. Isso explica por que é a Lua quem orbita ao redor da Terra e não ao contrário.
3. A Terra também sente uma aceleração devido à força gravitacional que exerce a Lua. Entretanto, essa aceleração é quase desprezível quando comparada com a que sente a Lua. A Terra então também "cai" em direção a Lua, mas como esta fica dando voltas, a Terra também acaba dando pequenas voltas ao redor de um centro.

</p>
