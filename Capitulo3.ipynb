## 1. Introdução
Provavelmente a maior lacuna dos alunos que ingressam na universidade é o fato de que não sabem isolar a incógnita de uma equação simples. Este capítulo está dedicado a isso. 

Isolar uma incógnita de uma equação não é mais que aplicar transformações na equação até conseguir nosso objetivo: deixar a incógnita à esquerda do igual e o resto á direita. Neste capítulo apresentaremos a teoria que fundamenta tais transformações, a geometria analítica, para com isso, aprender a isolar qualquer variável.

## 2. Fundamentos da geometria analítica
Dos arcabouços matemáticos introduzidos até agora, dois deles foram especialmente importantes: as funções e as equações, as quais, por exemplo, emergiam na hora de calcular raízes de uma função. Dedicamos um capítulo inteiro a analisar funções e todas suas propriedades, mas não analisamos tanto assim as equações. Aprofundemos então no significado de uma equação.

Usemos nossa querida função: $f(x)=2x+1$, da qual extraímos suas raízes. Para isso, como sabemos, devemos resolver a equação $2x+1=0$, sendo $x=-1/2$ a solução.

Sabemos efetivamente que $2x+1=0$ é uma equação e sabemos propriedades analíticas da mesma: equação linear ou polinômio de grau um, o qual deve ter uma solução segundo o Teorema Fundamental da Álgebra. Mas, que significa realmente essa equação do ponto de vista geométrico?

Podemos entender essa equação como se fosse uma igualdade de funções. Desta maneira, chamemos $g(x)$ à parte que fica à direita do igual e $h(x)$ à parte que fica à direita:
> $f(x)=2x+1$
>
> $g(x)=0$

e representemos tais funções num gráfico.

```python
# importing the required module
from turtle import color
import matplotlib.pyplot as plt

# x axis values
x_gx = [-2,-1,-0.5,0,1,2,3]
# corresponding y axis values
y_gx = [-3,-1,0,1,3,5,7]

# plotting the points
plt.plot(x_gx, y_gx, marker='')

# highlighting the solutions
plt.plot(x_gx[2], y_gx[2], marker='o', color='red')

# x axis values
x_hx = [-6,6]
# corresponding y axis values
y_hx = [0,0]

# plotting the points
plt.plot(x_hx, y_hx, marker='', color='blue')

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-2, 4)
plt.ylim(-2, 6)

plt.grid(linestyle = '--', linewidth = 0.5)

label = 'Raiz'

plt.text(x_gx[2], y_gx[2], label, ha = 'right', va = 'bottom', size = 12, color = 'red')

xorigem_ver = [0, 0]
yorigem_ver = [-2, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-2, 6]
yorigem_hor = [0, 0]
plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

plt.subplot(111).annotate("g(x)", xy=(3, 3), xytext=(2.2, 5))
plt.subplot(111).annotate("h(x)", xy=(3, 3), xytext=(3.5, 0.2))

# function to show the plot
plt.show()
```

Como vemos no gráfico, a raiz da equação $2x+1=0$ pode ser entendida também como o ponto de corte da reta $g(x)=2x+1$ com a reta $h(x)=0$, ou seja, o valor da coordenada $x$ onde as duas retas se atravessam.

Apliquemos agora uma transformação na equação. Por exemplo, somemos 2 unidades
> $(2x+1=0)+2$

Essa transformação deve ser aplicada em toda a equação, tanto à esquerda do igual quanto à direita:
> $2x+1+2=0+2$

Ou seja,
> $2x+3=2$

Podemos, a partir daqui, supor que tal equação está formada por 2 novas funções, chamemos-as de $f'(x)$ e $g'(x)$:
> $f'(x)=2x+3$
\
$g'(x)=2$

E, finalmente, representemos tais funções no gráfico anterior:

```python
# importing the required module
from turtle import color
import matplotlib.pyplot as plt

# x axis values
x_gx = [-2,-1,-0.5,0,1,2,3]
# corresponding y axis values
y_gx = [-3,-1,0,1,3,5,7]

# plotting the points
plt.plot(x_gx, y_gx, marker='', color='blue')

# highlighting the solutions
plt.plot(x_gx[2], y_gx[2], marker='o', color='red')

# x axis values
x_gx_linha = [-2,-1,-0.5,0,1,2,3]
# corresponding y axis values
y_gx_linha = [-1,1,2,3,5,7,9]

# plotting the points
plt.plot(x_gx_linha, y_gx_linha, marker='', color='orange')

# x axis values
x_hx = [-6,6]
# corresponding y axis values
y_hx = [0,0]

# plotting the points
plt.plot(x_hx, y_hx, marker='', color='blue')

# x axis values
x_hx_linha = [-6,6]
# corresponding y axis values
y_hx_linha = [2,2]

# plotting the points
plt.plot(x_hx_linha, y_hx_linha, marker='', color='orange')

# naming the x axis
plt.xlabel('Eixo X', loc='right')
# naming the y axis
plt.ylabel('Eixo Y')

# giving a title to my graph
plt.title('')

#axis limits
plt.xlim(-2, 4)
plt.ylim(-2, 6)

plt.grid(linestyle = '--', linewidth = 0.5)

label = 'Raiz'

plt.text(x_gx[2], y_gx[2], label, ha = 'right', va = 'bottom', size = 12, color = 'red')

xorigem_ver = [0, 0]
yorigem_ver = [-2, 6]
plt.plot(xorigem_ver, yorigem_ver, marker='', color='green', linestyle="--")

xorigem_hor = [-2, 6]
yorigem_hor = [0, 0]
#plt.plot(xorigem_hor, yorigem_hor, marker='', color='green', linestyle="--")

plt.subplot(111).annotate("g(x)", xy=(3, 3), xytext=(2.2, 5))
plt.subplot(111).annotate("g(x)'", xy=(3, 3), xytext=(1.2, 5))
plt.subplot(111).annotate("h(x)", xy=(3, 3), xytext=(3.5, 0.2))
plt.subplot(111).annotate("h'(x)", xy=(3, 3), xytext=(3.5, 2.2))

xraiz_ver = [-0.5, -0.5, -0.5]
yraiz_ver = [0, 1, 2]
plt.plot(xraiz_ver, yraiz_ver, marker='', color='red', linestyle="--")

# function to show the plot
plt.show()
```

Observemos detalhadamente, do ponto de vista geométrico, o resultado de tal transformação. Após ter somado 2 unidades, a primeira função, a função $g(x)$foi transformada em $g'(x)$ e, como resultado, esta última foi deslocada 2 unidades para cima. Além disso, a segunda função, a função $h(x)$ foi transformada em $h'(x)$ e, com resultado, esta última foi também deslocada 2 unidades para cima.

Esse foi o efeito de ter transformado a equação $2x+1=0$. Mas observe-se que, a pesar de tal transformação, **o valor da coordenada $x$ onde as funções se encontram continua sendo o mesmo**. De fato, a raiz da equação transformada ($2x+3=2$) volta a ser $x=-1/2$. Isto é assim porque as equações $2x+1=0$ e $2x+3=2$ são realmente as mesmas equações. De fato, através de uma conta básica podemos simplificar $2x+3=2$ para $2x+1=0$.

Pode parecer que todo o anterior só é valido para nosso exemplo, mas realmente não importa a transformação feita numa equação, que chegaremos à mesma conclusão: **As raízes de uma equação não mudam quando uma transformação é aplicada à mesma**.
