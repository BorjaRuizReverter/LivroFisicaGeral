# CAPÍTULO 10. Hidrostática

## 1. Introdução
Neste capítulo nos adentramos no estudo da Hidrostática. Hidrostática literalmente significa "água parada". Entretanto, a hidrostática envolve o estudo de qualquer líquido, não apenas água, desde que este esteja em equilíbrio mecânico. As grandezas físicas que aqui definiremos ainda se estendem para o estudo dos gases. Não assim com todos os teoremas que desenvolveremos. Para resolver tal ambiguidade explicitaremos quando uma ferramenta ou teorema é exclusiva de líquidos e quando é geral para fluidos, entendendo-se por líquidos e gases.

Toda a Física vista até agora consistiu em simplificar ao máximo a abordagem. Essa é maneira correta de proceder, como primeira aproximação, a um problema que, na vida real, já é complexo por se só. Essa simplificação consistia em aproximar o corpo por um ponto: a partícula, e em isolar ela do resto do Universo. No estudo dos fluidos essa abordagem não é mais válida. Um fluido é composto por um número enorme de partículas, as quais interagem umas com as outras através colisões, tipicamente em gases, ou através de ligações de Van de Waals, como acontece na água. Um fluido é um coletivo, que não pode ser desmembrado em entes individuais. E sendo assim, não resta outra que mudar o enfoque.

Como sempre, toda vez que mudamos o ponto de vista temos um *trade-off*, o qual tipicamente implica introduzir novas grandezas físicas. Neste caso, devemos definir novas grandezas físicas que embutam o fato de que o fluido é algo extenso e não mais um ponto isolado. Por exemplo, durante toda a análise da Dinâmica, usamos o conceito de massa, como sendo algo associado à partícula. Em fluidos, porém, usaremos uma grandeza física análoga que será a **densidade**, pois esta inclui a volume que um fluido ocupa no espaço, evitando assim o conceito de partícula. Também, toda a Dinâmica orbitava ao redor do conceito de força aplicada na partícula. Entretanto, não faz nenhum sentido aplicar uma força a um ponto de um fluido. O que faz sentido é aplicar uma força sobre alguma área do fluido, abrangendo um grande número de partículas, o qual implica em exercer uma **pressão** sobre o fluido.

Desta maneira, estamos realizando uma espécie de transformação de grandezas físicas. Ao invés de usar os conceitos de massa e força, os quais não fazem mais sentido no estudo dos fluidos, usamos os conceitos de densidade e pressão, os quais envolvem volumes e áreas, dando assim um enfase ao conceito de fluido como o que realmente é: um ente extenso. Como veremos, muitas das equações vistas até agora em Dinâmica resultam ser as mesmas para fluidos apenas realizando essa transformação.

## 2. Densidade
Com foi indicado na introdução, temos um novo desafio: descrever os fluidos. Mas não temos as ferramentas adequadas para isso. Em Física estas ferramentas são as grandezas físicas. E, sendo assim, precisamos criar novas ferramentas a partir de definições. Definições que sejam úteis para nossos propósitos.

Uma grandeza física necessária para caracterizar um fluido é sua **densidade**. A densidade ($\rho$) é definida como
$$\boxed{\tag{*}\rho = \frac{m}{V}}$$
, onde $m$ é a massa do fluido e $V$ é o volume que ele ocupa. Como, no Sistema Internacional, as unidades da massa são os quilogramas e as unidades do volume os metros cúbicos, a densidade tem unidades de $kg/m^3$,

Essas unidades, porém, são relativamente grandes. Por isso, é comum usar outras menores como gramas sobre litro ($g/L$) ou $g/cm^3$. É importante neste ponto lembrar dos fatores de conversão para poder transformar de umas para outras se precisar.

## 3. Pressão
Outra grandeza física importante para estudar fluidos é a **pressão**. Da maneira geral, não só para o estudo particular dos fluidos, definimos a pressão como:

$$\boxed{\tag{*}p \equiv \frac{F}{A}}$$
, onde $F$ é a força aplicada numa área $A$ específica. De maneira geral, uma pressão é uma força transmitida através de uma área (<mark>figura *</mark>).

<img src="./figures/pressure.png" alt="alt text" title="image Title" width="30%"/>

No Sistema Internacional as unidades da força são os $N$, ou equivalentemente, $kg\cdot m/s^2$ e as unidades da área são $m^2$. Sendo assim, as unidades da pressão serão $N/m^2$, ou equivalentemente, $kg/m\cdot s^2$, os quais se bautizam como Pascais (P), em honor ao físico frances Blaise Pascal.

Existem, porém, inúmeras unidades alternativas para a pressão. A <mark>tabela *</mark> resume algumas delas.

| Unidade  | Abreviatura   | Equivalencia |
| :--------------: | :-------: | :-------: |
| Bares | $bar$ | $1bar = 100000Pa$|
| Atmosferas | $atm$ | $1atm = 101325Pa$|
| Torricelli | $torr$ | $1torr = 1mmHg$|
| Milímetros de mercurio | $mmHg$ | $1mmHg \approx 133Pa$|
| Libras por polegada quadrada (psi) | $psi$ | $1psi \approx 6895Pa$|

Algumas destas unidades se usam frequentemente no cotidiano. A técnica dos fatores de conversão resulta útil para transformar entre elas.

<b>As lâminas de um patin de gelo ocupam aproximadamente uma área de $4 cm^2$. Enquanto que a área da sola de um sapatenis ocupa uma área aproximada de $400 cm^2$. Se uma patinadora $60 kg$ percorre uma pista de gelo, calcule qual será a pressão que o chão de gelo experimenta quando ela calza os sapatenis e quando vai subida nos patins de gelo. Explique por que os patins são capazes de furar o pista de gelo.</b>

<p>

Usando os sapatenis, supondo que os dos pés estão no chão ao mesmo tempo, a pressão que o chão sente seria:

</p>

$$p = \frac{m\cdot g}{2\cdot A_{sapatenis}} = \frac{60 kg \cdot 10 \frac{m}{s^2}}{2\cdot 4\cdot 10^2 cm^2(\frac{1m}{10^2cm})^2} = \frac{6}{8}\frac{10^2}{10^{-2}}\frac{kg\frac{m}{s^2}}{m^2} = \frac{3}{4}10^4\frac{kg}{m\cdot s^2} = 7.5kPa$$

<p>

Já usando os patins de gelo, supondo que os dos pés estão no chão ao mesmo tempo, a pressão que o chão sente seria:

</p>

$$p = \frac{m\cdot g}{2\cdot A_{patins}} = \frac{60 kg \cdot 10 \frac{m}{s^2}}{2\cdot 5 cm^2(\frac{1m}{10^2cm})^2} = 6\cdot \frac{10^2}{10^{-3}}\frac{kg\frac{m}{s^2}}{m^2} = 6\cdot 10^5\frac{kg}{m\cdot s^2} = 600 kPa$$

<p>
A pressão exercida no chão pelos patins de gelo é ao redor de 100 vezes maior que a exercida usando uns sapatenis. Isso explica porque o gelo é furado ao andar com patins de gelo.

## 4. Pressão atmosférica
A ideia de que a pressão é a força transmitida através de uma área é clara quando a pressão é exercida por um corpo sólido contra outro similar, mas não tanto quando consideramos fluidos. Isso é devido a que um fluido se caracteriza por estar composto por muitas partículas livres ou quase livres e, sendo assim, não resulta tão claro quem exerce a força. De fato, quando se trata de um fluido, tal força surge como resultante de inúmeras colisões atômicas ou moleculares das partículas do fluido contra o sólido em questão. Por exemplo, um gás, ou um líquido, exerce uma pressão sobre as paredes de um recipiente devido às inúmeras colisões das partículas que o compõem contra a superfície do recipiente.

Da mesma maneira, a atmosfera, composta por principalmente por nitrogênio e oxigênio em estado molecular, exerce uma pressão sobre qualquer objeto no seno dela devido a colisões moleculares. Entretanto, a diferença de um fluido dentro de um recipiente cotidiano, a densidade da atmosfera diminui com a altura devido à menor força de gravidade. Sendo assim, a medida que a altura aumenta, a densidade cai, e se massa de partículas por unidade de volume diminui, espera-se que também o faça o número de colisões por metro quadrado de área (<mark>figura *</mark>).

## 5. Experiência de Torricelli
Ao redor de 1640 o físico italiano Evangelista Torricelli realizou uma série de experimentos com o propósito de mensurar a pressão atmosférica. Para isso, Torricelli primeiro encheu de mercúrio líquido um tubo de 1 metro de largura e depois virou o tubo cabeça para baixo, colocando-o sobre um balde. Ao fazer isso, o mercúrio vazava do tubo, mas nunca por completo, pois a maior parte ficava dentro do tubo. De fato, a coluna de mercúrio se mantinha sempre em 760 mm de largura.

<img src="./figures/exp-torricelli.jpg" alt="alt text" title="image Title" width="30%"/>

Torricelli pensou que o tubo não se esvaziava por completo devido a estar em equilíbrio dinâmico. A ideia é que isso acontecia devido a que o peso da atmosfera sobre a área ocupada pelo mercúrio vertido no balde equilibrava o peso da coluna de mercúrio sobre a boca do tubo, evitando assim maiores vazamentos. Torricelli imaginava sua experiência como um balanço em equilíbrio, no qual a atmosfera puxa de um lado e o peso da coluna de mercúrio por outro.


## 6. Equação fundamental da hidrostática
Uma das questões mais importantes a ser resolvida em Hidrostática é encontrar uma relação da pressão que exerce um fluido em função da profundidade. No caso da atmosfera a densidade varia muito com altura e desenvolver tal funcionalidade não é imediata, pois precisa realizar um cálculo integral. Entretanto, a maioria dos líquidos satisfaz a chamada condição de **incomprensibilidade**, o que implica que a densidade do fluido não se altera dentro de um rango considerável de pressões. Esta condição facilita muito o desenvolvimento de tal expressão, pois a todos os efeitos podemos considerar que a densidade é aproximadamente constante com a profundidade. Usaremos a seguir esta condição para estabelecer como muda a pressão com a profundidade no caso dos líquidos.

Para encontrar essa expressão escolhamos um elemento cilíndrico de um líquido qualquer de densidade $\rho$ no seno do campo gravitacional terrestre dado por $g$. Esse elemento possui um volume $V$, o qual pode ser explicitado pelo produto da área da base $A$ vezes a largura $h$ do cilindro, e também tem um peso $\vec{P}$.

Se apenas atuasse a força peso, o elemento estaria em queda livre. Entretanto, como ele está em equilíbrio dinâmico outras forças devem intervir. Efetivamente, por estar dentro do fluido, as hipotéticas paredes desse elemento estão sometidas a forças, dadas pelas colisões das partículas do líquido. De fato, as partículas colidem com as "paredes" da face de baixo, as quais poderíamos somar e sintetizar na força resultante $\vec{F}_2$ apontando para cima, e também nas "paredes" da face de cima, as quais poderíamos resumir na força $\vec{F}_1$ apontando para baixo. E já que o elemento está equilibro, aplicando a segunda Lei de Newton no eixo Y, deve acontecer que:
$$F_2 = F_1 + P$$

As partículas do líquido também colidem com as paredes laterais, mas por simetria e também devido ao equilibrio dinâmico na horizontal, as forças laterais se equilibram.

## 7. Vasos comunicantes
Uma das consequências da Equação Fundamental da Hidrostática é o fato de que a pressão apenas depende da profundidade. Isso implica que a pressão num ponto do fluido não depende da forma do recipiente. Podemos usar recicipientes de geometrias diversas (<mark>figura *</mark>) e comprovar que a pressão a uma mesma profundidade $h$ será a mesma em todos. 

Essa propriedade deu lugar aos chamados **vasos comunicantes**. Isto é, recipientes em forma de U, que se caracterizam por possuirem a mesma pressão às mesmas profundidades.
